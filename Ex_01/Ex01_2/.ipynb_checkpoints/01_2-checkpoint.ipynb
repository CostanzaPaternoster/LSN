# Ez. 1.2 - Verifica del teorema del limite centrale (TLC)

Nella prima parte di questo esercizio è stata ampliata la classe Random con l'aggiunta di nuove funzioni per il campionamento delle distribuzioni di probabilità:
- esponenziale decrescente $p(x) = \lambda \exp(-\lambda x)$, $x\in [0;+\infty]$
- di Cauchy-Lorentz $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$

In seguito, sono state utilizzate queste due distribuzioni e la distribuzione uniforme sull'intervallo $[0,1)$ per verificare la validità e i limiti di applicabilità del TLC.

Per fare ciò, sono state realizzate simulazioni MC per valutare la distribuzione della variabile casuale $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (per $N=1, 2, 10, 100$) - dove le $x_i$ sono state generate secondo una delle tre distribuzioni sopracitate. Gli istogrammi sono stati costruiti con un totale di $10^4$ realizzazioni di $S_N$. 
Il TLC asserisce che se le RV $\{x_i\}$ sono IID con media $\mu$ e varianza $\sigma^2$ finita, la RV $S_N$ avrà media $\mu$ e varianza $\sigma^2/N$ e sarà distribuita secondo una normale $\mathscr{N}(\mu,\sigma/\sqrt{N})$. Tutto ciò a patto di considerare un numero $N$ sufficientemnte grande, e quindi ci si aspetta che al crescere di $N$ la distribuzione delle $S_N$ approssimi sempre di più quella di una normale. In questa forma, il TLC è valido per le distribuzioni uniforme e esponenziale, ma non per la lorentziana - che ha varianza non definita. La versione generalizzata del TLC afferma che la distribuzione limite di $S_N$ di una lorentziana rimane una lorentziana, ed è ciò che è stato verificato in questo esercizio. 

Di seguito si riportano 4 istogrammi per le 3 distribuzioni di probabilità e un grafico che mostra il fit dell'istogramma a $N=100$ con l'appropriata distribuzione di probabilità limite attesa. Gli istogrammi sono stati opportunamente normalizzati.

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
import decimal

# funzione gaussiana
def Gauss(x,a,b,c):
    return a * np.exp(-b*(x-c)**2)

# funzione lorentziana
def Lorentz(x,a,b):
    return (1/math.pi)*b/((x-a)**2+b**2)
```

### Distribuzione uniforme $p(x)=1$ $ x\in[0,1)$
- $\mu=1/2$
- $\sigma=\sqrt{1/12}$

NB: per poter essere confrontati, gli istogrammi sono stati riscalati della quantità $\sigma/\sqrt{N}$ e traslati del valor medio.

```python
mean=0.5
sigma=np.sqrt(1/12)

# numero bin e range istogramma
n_bins = 50
width=5

# Caricamento da file delle realizzazioni delle S_N per i 4 valori di N (1,2,10,100)
x1, x2, x10, x100 = np.loadtxt("Output/Uniform.txt", usecols=(0,1,2,3), delimiter = ' ', unpack = 'true')

# Normalizzazione istogrammi
x1=(x1-mean)/sigma
x2= (x2-mean)*np.sqrt(2)/sigma 
x10=(x10-mean)*np.sqrt(10)/sigma
x100=(x100-mean)*np.sqrt(100)/sigma

X = [x1,x2,x10,x100]

title = ['N=1','N=2','N=10','N=100']
color = ['g','b','r','y']

# Plot istogrammi
f = plt.figure(figsize=(12,10))

for i in range(4):
    g = f.add_subplot(2,2,i+1)
    plt.xlabel('x')
    plt.title('Istogramma normalizzato '+title[i])
    n, bins, patches = g.hist(X[i], n_bins, density=True, facecolor=color[i], range=(-width,width))
    plt.grid(True)

plt.show()

N=len(n)
x=np.linspace(-width,width,N,endpoint=True)

# plot funzione che approssima istogramma
plt.plot(x,n, label = 'data')

# fit con gaussiana
p_opt, p_cov = curve_fit(Gauss, x, n, bounds=([0,0,-1],[2,3,+3]))
y_fit = Gauss(x,p_opt[0],p_opt[1],p_opt[2])
plt.plot(x,y_fit,label='fit') # plotting fitted function
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Fit con gaussiana')
plt.legend()
plt.grid(True)
plt.show()

print("Parametri fit:\n",round(p_opt[0],3),round(p_opt[1],3),round(p_opt[2],3),'\n')
print("Parametri attesi per distribuzione normale:\n",round(np.sqrt(1/2/math.pi),3),1/2,0)
```

### Distribuzione esponenziale $p(x)=e^{-x}$ $x\in [0;+\infty]$
- $\mu=1$
- $\sigma=1$

NB: per poter essere confrontati, gli istogrammi sono stati riscalati della quantità $\sigma/\sqrt{N}$ e traslati del valor medio.

```python
mean=1
sigma=1

# numero bin e range istogramma
n_bins = 50
width=5

# Caricamento da file delle realizzazioni delle S_N per i 4 valori di N (1,2,10,100)
x1, x2, x10, x100 = np.loadtxt("Output/Exp.txt", usecols=(0,1,2,3), delimiter = ' ', unpack = 'true')

# Normalizzazione istogrammi
x1=(x1-mean)/sigma
x2= (x2-mean)*np.sqrt(2)/sigma 
x10=(x10-mean)*np.sqrt(10)/sigma
x100=(x100-mean)*np.sqrt(100)/sigma

X = [x1,x2,x10,x100]

title = ['N=1','N=2','N=10','N=100']
color = ['g','b','r','y']

# Plot istogrammi
f = plt.figure(figsize=(12,10))

for i in range(4):
    g = f.add_subplot(2,2,i+1)
    plt.xlabel('x')
    plt.title('Istogramma normalizzato '+title[i])
    n, bins, patches = g.hist(X[i], n_bins, density=True, facecolor=color[i], range=(-width,width))
    plt.grid(True)

plt.show()

N=len(n)
x=np.linspace(-width,width,N,endpoint=True)

# plot funzione che approssima istogramma
plt.plot(x,n, label = 'data')

# fit con gaussiana
p_opt, p_cov = curve_fit(Gauss, x, n, bounds=([0,0,-1],[2,3,+3]))
y_fit = Gauss(x,p_opt[0],p_opt[1],p_opt[2])
plt.plot(x,y_fit,label='fit') # plotting fitted function
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Fit con gaussiana')
plt.legend()
plt.grid(True)
plt.show()

print("Parametri fit:\n",round(p_opt[0],3),round(p_opt[1],3),round(p_opt[2],3),'\n')
print("Parametri attesi per distribuzione normale:\n",round(np.sqrt(1/2/math.pi),3),1/2,0)
```

### Distribuzione lorentziana $p(x)=\frac{1}{\pi}\frac{1}{x^2+1}$, $x\in [-\infty;+\infty]$
- $\mu=0$
- $\sigma$ non definita

```python
mean=0

# numero bin e range istogramma
n_bins = 50
width=10

# Caricamento da file delle realizzazioni delle S_N per i 4 valori di N (1,2,10,100)
x1, x2, x10, x100 = np.loadtxt("Output/Lorentz.txt", usecols=(0,1,2,3), delimiter = ' ', unpack = 'true')

X = [x1,x2,x10,x100]

title = ['N=1','N=2','N=10','N=100']
color = ['g','b','r','y']

# Plot istogrammi
f = plt.figure(figsize=(12,10))

for i in range(4):
    g = f.add_subplot(2,2,i+1)
    plt.xlabel('x')
    plt.title('Istogramma normalizzato '+title[i])
    n, bins, patches = g.hist(X[i], n_bins, density=True, facecolor=color[i], range=(-width,width))
    plt.grid(True)

plt.show()

N=len(n)
x=np.linspace(-width,width,N,endpoint=True)

# plot funzione che approssima istogramma
plt.plot(x,n, label = 'data')

# fit con lorentziana
p_opt, p_cov = curve_fit(Lorentz, x, n)
y_fit = Lorentz(x,p_opt[0],p_opt[1])
plt.plot(x,y_fit,label='fit') 
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Fit con lorentziana')
plt.legend()
plt.grid(True)

print("Parametri fit:\n",round(p_opt[0],3),round(p_opt[1],3),'\n')
print("Parametri attesi per distribuzione lorentziana:\n",0,1)
```

Nei primi due casi (distribuzioni uniforme ed esponenziale) è ben visibile il progressivo avvicinamento della distribuzione delle $S_N$ alla funzione gaussiana all'aumentare di $N$. I parametri ottenuti dai fit degli istogrammi a $N=100$ con la funzione gaussiana $\mathscr{N}(\mu,\sigma/\sqrt{N})$ sono in ottimo accordo con i valori teorici attesi.

Nell'ultimo caso, invece, la variabile aleatoria $S_N$ non modifica la sua distribuzione all'aumentare di $N$, come si può osservare anche dal fit effettuato con funzione lorentziana. Anche questo comportamento è in accordo con il TLC: non avendo momento secondo finito, la distribuzione lorentziana non converge alla distribuzione normale e il suo bacino di attrazione (nello spazioe delle distribuzioni di probabilità) è la lorentziana stessa.
