# Es. 1.3  - Simulazione dell'esperimento di Buffon: stima di $\pi$ con metodi MC


In questo esercizio è stato simulato l'esperimento dell'ago di Buffon, un classico esempio di metodo MC volto ad ottenere una stima del valore di $\pi$. Nell'esperimento, un ago di lunghezza $L$ viene lanciato in maniera casuale su un piano diviso da righe parallele distanti $d$ ($d>L$) tra loro. Si può dimostrare che la probabilità $P$ che l'ago intersechi una linea è $P=\frac{2L}{\pi d}$. Secondo una definizione frequentista, la probabilità può essere espressa come rapporto tra esiti favorevoli ed esiti totali e quindi $P=N_{int}/N_{tot}$, dove $N_{tot}$ è il numero totale di lanci e $N_{int}$ il numero di lanci in cui l'ago interseca una linea. Eguagliando le due definizioni di probabilità è quindi possibile ricavare un'espressione per $\pi$:

$$\pi\simeq\frac{2L N_{tot}}{d N_{int}}, N_{tot}\rightarrow\infty $$

Per simulare l'esperimento, quindi, sono stati effettuati $N_{tot}=10^5$ lanci per ottenere una singola stima di $\pi$. La procedura è stata ripetuta $N=100$ volte per poter mediare su un numero maggiore di stime di $\pi$ e per  calcolare l'errore statistico.

Di seguito si riporta il grafico della stima di $\pi$ e relativs incertezza statistica in funzione del numero di blocchi.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import decimal

# caricamento da file della stima di pi-greco e dell'errore sulla stima di pi-greco in funzione del n di blocchi 
ave_pi,err_pi = np.loadtxt("Output/Pi.txt", usecols=(0,1), delimiter = ' ', unpack = 'true')

N = len(ave_pi)
x = np.arange(N)

# grafico stima di pi in funzione del numero di blocchi
plt.errorbar(x,ave_pi,yerr=err_pi)
plt.xlabel('# blocks')
plt.ylabel('$\pi$')
plt.title('Stima di $\pi$ in funzione del numero di blocchi',size=15)
plt.grid(True)
plt.show()

print("Miglior stima di pi-greco: ",round(ave_pi[N-1],3),"+/-",round(err_pi[N-1],3))
print("Valore vero di pi-greco: ",math.pi)
```

Il valore di $\pi$ ottenuto dalla simulazione è in buon accordo con il valore vero.
