## Perfomance del modello DNN per diversi metodi di ottimizzazione

In questo esercizio abbiamo testato le performance di una DNN (Deep Neural Network) nel problema della classificazione di immagini rappresentanti cifre, un classico nell'introduzione alle reti neurali, al variare del metodo di ottimizzaizone scelto nella fase di allenamento della rete.

Vediamo nel dettaglio il problema.

### Struttura della DNN

La struttura delle rete è quella presentata nella consegna dell'esercizio $LSN\text{_}Exercise\text{_}12$, e per non allungare inutilmente il codice abbiamo utilizzato direttamente quel jupyter notebook per creare e allenare le reti, salvando i risultati e alcuni file .log contenenti la "training history". In particolare la struttura della DNN è la seguente:
- un layer Dense composto da 400 neuroni con attivazione di tipo ReLu
- un layer Dense di 100 neuroni con attivazione ReLu
- un'ultimo layer Dense di 10 neuroni connesso con un Dropout rate ~ 0.5 al layer precedente, e attivazione di tipo SoftMax per la classificazione 

Una rete di questo tipo ha un numero di parametri pari a $355110$!

```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam
from keras.datasets import mnist
```

### Metodi di ottimizzazione

Abbiamo allenato la rete con la struttura spiegata sopra con 3 diversi metodi di ottimizzazione.

#### Gradient descent stocastico SGD
La scelta standard è quella del metodo $\textit{SGD}$: è un miglioramento del metodo classico GD, il quale diventa troppo dispendioso per dataset grandi, che prevede di valutare il gradiente non più rispetto all'intero dataset, bensì rispetto a un solo elemento alla volta (o a un $\textit{mini-batch}$), mescolando ogni volta il dataset per garantire un corretto funzionamento.

#### Adapted gradient descent Adagrad
Come suggerisce il nome, l'$\textit{Adagrad}$ adatta il metodo di gradient descent ai singoli parametri del modello: in altre parole ogni parametro viene qui modificato con un $\textit{learning rate}$ differente, che sarà grande per parametri che vengono aggiornati raramente (features meno comuni) e piccolo per parametri che vengono aggiornati spesso (features più comuni). Il vantaggio di Adagrad su SGD è evidente, ma Adagrad ha anche svantaggi: nella sua implementazione infatti è previsto l'accumulo dei gradienti della funzione costo rispetto a ogni parametro $\textit{ad ogni istante di tempo}$. Questo ha due conseguenze: memoria occupata per lo storing di tutti questi valori e (sempre a causa della sua implementazione) un learning rate che tende a 0 col passare del tempo.

#### Adadelta
Questi due problemi sono risolti dal metodo $\textit{Adadelta}$. Invece che salvare il valore dei gradienti a tutti gli step precedenti, viene utilizzata solo una finestra $\mathcal w $ di valori precedenti: meglio ancora, la finestra viene usata solo per calcolare una media dei valori degli ultimi $\mathcal w$ gradienti. Questa accortezza risolve enyrambi i problemi del metodo Adagrad.

### Performance dei 3 modelli di ottimizzazione

Studiamo in questa fase la performance di una DNN  strutturata come spiegato sopra e ottimizzata con i 3 precedenti metodi. 

La performance è valutata sui file del database MNIST, contenente 60000 immagini di cifre scritte a mano da 250 persone fra impiegati dell'US Census Bureau e studenti di scuole superiori: queste immagini sono divise in due set da 50000 e 10000 immagini, utilizzate rispettivamente per il training della rete e per il successivo test su dati nuovi (fase di $\textit{validation}$). 

Come anticipato, ci limitiamo qui a caricare le reti, allenate precedentemente, per valutarne la performance sui dati di validation.

```python
# Loading previously trained DNN
model_old1 = tf.keras.models.load_model(filepath='Saved/model_DNN_Sgd')
model_old2 = tf.keras.models.load_model(filepath='Saved/model_DNN_Adagrad')
model_old3 = tf.keras.models.load_model(filepath='Saved/model_DNN_Adadelta')
```

```python
# Loading MINST data
# Input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# Output
num_classes = 10 # 10 digits

# Data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

# Reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

# Cast floats to single precesion
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# Rescale data in interval [0,1]
X_train /= 255
X_test /= 255

# Convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
```

```python
# Evaluate model (it takes some seconds)
score1_train = model_old1.evaluate(X_train, Y_train, verbose=0)
score2_train = model_old2.evaluate(X_train, Y_train, verbose=0)
score3_train = model_old3.evaluate(X_train, Y_train, verbose=0)
score1_test = model_old1.evaluate(X_test, Y_test, verbose=0)
score2_test = model_old2.evaluate(X_test, Y_test, verbose=0)
score3_test = model_old3.evaluate(X_test, Y_test, verbose=0)
```

```python
# Print performance
print('\nSGD MODEL')
print('Train loss:', score1_train[0])
print('Train accuracy:', score1_train[1])
print('Test loss:', score1_test[0])
print('Test accuracy:', score1_test[1])
print('\nADAGRAD MODEL')
print('Train loss:', score2_train[0])
print('Train accuracy:', score2_train[1])
print('Test loss:', score2_test[0])
print('Test accuracy:', score2_test[1])
print('\nADADELTA MODEL')
print('Train loss:', score3_train[0])
print('Train accuracy:', score3_train[1])
print('Test loss:', score3_test[0])
print('Test accuracy:', score3_test[1])
```

L'$\textit{accuracy}$ è una misura della percentuale di classificazioni corrette effettuate dall rete, mentre la $loss$ è una funzione utilizzata per caratterizzare l'errore commesso dalla rete. Sebbene l'informazione data da queste due quantità sia simile, è importante sottolineare che il monitoraggio di entrambe le quantità può essere utile, ad esempio per controllare la comparsa del cosidetto $\textit{overfitting}$.

Nel nostro caso, trattandosi di una rete che deve effettuare una classificazione, la funzione loss che si è utilizzata è la $\textit{categorical cross-entropy}$

$$
\mathcal{S}[y^{lab}; y^{out}]=\frac{1}{N}\sum_{i=1}^{N}\big(y_i^{out}\cdot\log y_i^{lab}\big) \;.
$$

Nei risultati riportati notiamo come i metodi Adagrad e Adadelta performino in maniera molto simile (accuracy ~98%), e migliore del metodo SGD (accuracy ~95%). In tutti i casi, come ci si aspetta, l'accuratezza sui dati di training è $\textit{maggiore}$ di quella sui dati di validation, essendo i primi quelli su cui la rete è stata allenata.

### Studio della training history

Ci soffermiamo ora sullo studio del training dei 3 modelli, per evidenziare eventuali differenze. Per fare questo sono stati salvati (tramite il $\textit{callback}$ $\textit{CSVLogger}$ di Keras) file .log contenenti informazioni temporali raccolte durante l'allenamento.

```python
text = ['Sgd', 'Adagrad', 'Adadelta']
f = plt.figure(figsize = (12, 12))

for i in range(3):
    # Load history of the training 
    history_read = pd.read_csv('Saved/training_'+ text[i]+'.log')# , engine='python'
    
    ax = f.add_subplot(3, 2, 2*i+1)
    
    # Plot accuracy on training and test data
    plt.plot(history_read['epoch'],history_read['acc'], label = 'train')
    plt.plot(history_read['epoch'], history_read['val_acc'], label = 'test' )
    
    # Add some useful information to the drawing  
    if (i == 0): 
        plt.title('Model Accuracy', fontweight = 'bold')
        plt.xlabel('epoch')
        plt.legend()
        ax.text(2, 0.85, text[i], style='italic', size = 15,
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})    
    else:
        ax.text(2, 0.94, text[i], style='italic', size = 15,
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})        
    
    ax = f.add_subplot(3, 2, 2*i+2)
   
    # Plot loss function for training and test data
    plt.plot(history_read['epoch'],history_read['loss'], label = 'train')
    plt.plot(history_read['epoch'], history_read['val_loss'], label = 'test' )
    
    # Again, add some information 
    if (i == 0): 
        plt.title('Model Loss', fontweight = 'bold')
        ax.text(2, 0.6, text[i], style='italic', size = 15,
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})    
    else:
        ax.text(2, 0.2, text[i], style='italic', size = 15,
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})    
    

```

Tutte le informazioni necessarie per interpretare le curve sono contenute nei primi due grafici: ancora una volta verifichiamo che le reti allenate con Adagrad e Adadelta convergono più velocemente a performance ottimali rispetto alla reteallenata con SGD.

Un aspetto interessante qui è che nella quasi totalità dei casi l'accuratezza di validation è $\textit{maggiore}$ dell'accuratezza di training, in contrasto con quanto notato sopra per le perfomance del modello. Questo fatto è dovuto alla presenza del $\textit{Dropout}$ nella rete: attivo in fase di training, quest'ultimo la costringe a "imparare più modi per arrivare ad uno stesso risultato", abbassandone inevitabilmente la performance; nella fase di validation (e quindi nel metodo $\textit{model.evaluate}$ chiamato sopra) invece il Dropout si disabilita, permettendo alla rete di agire in tutta la sua potenza.

Notiamo infine che nella rete allenata con Adadelta dopo poco più di due epoche di training avviene il "crossing" tra le performance sui dati di training e su quelli di validation. Detto in altre parole, la rete continua a migliorare sui dati di training (nonstante il Dropout), senza mostrare un miglioramento analogo sui dati di validation.
