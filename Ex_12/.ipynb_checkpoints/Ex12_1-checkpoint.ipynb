$[y^{lab}; y^{out}]=\frac{1}{N}\sum_{i=1}^{N}\big(y_i^{out}\cdot\log y_i^{lab}\big)$ 

Per testare la performance del modello, invece, è stata utilizzata l' 'accuracy', una misura della percentuale di classificazioni corrette effettuate dalla rete.

## Ex 12.1 - Perfomance del modello DNN per diversi metodi di ottimizzazione

In questo esercizio sono state allenate e testate le performance di tre Deep Neural Networks ottenute mediante tre metodi di ottimizzazione del tipo "gradient descent":

- Mini-batch Gradient Descent
- Adagrad
- Adadelta

I "gradient-descent" sono una classe di algoritmi ampiamente utilizzati nell'ottimizzazione delle reti neurali. Questi algoritmi permettono di minimizzare una funzione $J(\theta)$ dei parametri $\theta E R^{d}$ del modello (la funzione di costo) aggiornando i parametri nella direzione opposta al gradiente $grad_{\theta}J(\theta)$ della funzione rispetto ai parametri. Il learning rate $\eta$ determina la dimensione dei passi nello spazio dei parametri.

#### BATCH/STOCHASTIC/MINI_BATCH GRADIENT DESCENT (BGD,SGD,MBGD)",
Tra i metodi "gradient-descent" più semplici vi sono i Batch, Stochastic e Mini-batch Gradient Descent. Queste tre varianti differiscono nella quantità di dati utilizzati per calcolare il gradiente della funzione di costo. A seconda della quantità di dati, si può scegliere se migliorare l'accuratezza nell'aggiornamento dei parametri o ottimizzare il tempo impiegato per eseguire un aggiornamento. 

Il metodo BGD descent calcola il gradiente della funzione di costo per l'intero dataset di training:

$\theta = \theta + \eta grad_{\theta}J(\theta)$

Poichè ad ogni singolo aggiornamento viene calcolato il gradiente per l'intero set di dati di training, il metodo BGD risulta molto lento. Questo metodo permette di convergere al minimo globale di una funzione d'errore convessa e al minimo locale nel caso di superfici non convesse.

Lo SGD, a differenza del BGD, effettua l'aggiornamento dei parametri per ciascun dato di training (coppia $(x_{i},y_{i})$:

$\theta = \theta + \eta grad_{\theta}J(\theta;x^{(i)};y^{(i)})$

In questo modo, questo algoritmo evita di eseguire calcoli che risulterebbero ridondanti per dataset molto grandi e, di conseguenza, il tempo necessario ad aggiornare i parametri diminuisce notevolmente. Inoltre, gli aggiornamenti dello SGD sono caratterizzati da una grande varianza, che permette alla funzione di costo di fluttuare molto di più rispetto al caso precedente. Questa caratteristica permette di eseguire un campionamento più completo dello spazio dei parametri dato che evita di rimanere incastrati in un minimo locale della funzione di costo. Di contro, i continui aggiornamenti rendono più complicata la convergenza al minimo esatto.

Il metodo MBGD è un ibrido fra i due sopracitati e permette di ottenere i vantaggi di entrambi: il set di dati di training viene diviso in sottoinsiemi (mini-batch) e l'aggiornamento dei parametri si ottiene calcolando il gradiente ciascun sottoinsieme. La grandezza di ciascun mini-batch viene scelta in base al numero di dati di training. 

$\theta = \theta + \eta grad_{\theta}J(\theta;x^{(i:i+n)};y^{(i:i+n)})$

#### ADAGRAD

Adagrad è un algoritmo di ottimizzazione del tipo "gradient descent" che adatta il learning rate $\eta$ a seconda dei parametri: esegue piccoli passi (basso learning rate) per parametri che vengono aggiornati di frequente e passi più grandi (alto learning rate) per parametri che vengono aggiornati con frequenza più bassa. In pratica, Adagard adatta gli aggiornamenti di ciascun parametro a seconda della sua importanza. Per questo motivo, si presta ad essere utilizzato quando si hanno pochi dati di trainig a disposizione.

$g_{t,i} := grad_{\theta}J(\theta_{i})$ (derivata parziale della f. di costo rispetto al parametro $i$-esimo al time step $t$)

Aggiornamento del parametro $i$-esimo:

$\theta_{i,t+1}= \theta_{i,t} + \eta_{i,t} g_{t,i}$

Il learning rate $\eta_{i,t}$ di ciascun parametro viene modificato ad ogni time step $t$ sulla base dei gradienti calcolati a tempi precedenti per lo stesso parametro:

$\theta_{i,t+1}= \theta_{i,t} + \frac{g_{t,i}}{\sqrt{G_{t,ii}+\epsilon}}$

$G_{t}$ è una matrice diagonale dove ciascun elemento diagonale $i,i$ è la somma dei quadrati dei gradienti - rispetto a $\theta_{i}$ calcolati ai time step precedenti. $\epsilon$ è un termine che evita la divisione per zero.

A causa dell'accumulo di gradienti (al quadrato) al denominatore, il learning rate diminuisce ad ogni time step: nei casi in cui questo diventi eccessivamente piccolo, l'algoritmo rischia di non essere più in grado di allenare la rete.

#### ADADELTA

Il problema legato ad Adagrad viene risolto con il metodo di ottimizzazione Adadelta: invece di accumulare tutti i quadrati dei gradienti precedenti, Adadelta limita l'accumulo ad un tot di time step precedenti. La somma dei gradienti precedenti, inoltre, non viene aggiornata di volta in volta (operazione che richiede tempo e memoria), ma viene definita in maniera ricorsiva applicando una funzione di decadimento (come un esponenziale decrescente) alla media quadratica $E[g^{2}]_{t}$ dei gradienti.

$E[g^{2}]_{t} = \gamma E[g^{2}]_{t-1} + (1 - \gamma)g^{2}_{t}$

$\theta_{t+1}= \theta_{t} + \frac{g_{t}}{\sqrt{E[g^{2}]_{t}+\epsilon}}$

### Esercizio:

Si è deciso di confrontare gli errori sui dati di training/test e l'accuratezza di tre reti neurali ottenute attraverso i 3 metodi di ottimizzazione precedentemente illustrati:
- model_DNN1: SGD (mini-batch)
- model_DNN2: Adagrad
- model_DNN3: Adadelta
I parametri sono stati mantenuti con i loro valori di default.

Dati:
- 60000 dati di training (input: vettore 1d di 28x28=784 float in [0,1], output: vettore 1d di 10 float in [0,1])
- 10000 dati di test (",")

Caratteristiche delle reti (DNN):
- 1 all-to-all relu layer con 400 neuroni
- 1 all-to-all relu layer con 100 neuroni
- 1 soft-max layer con 10 neuroni di output

Caratteristiche dei metodi di training e validazione:
- funzione di costo: categorical cross entropy
- valutazione performance: accuracy
- metodo di ottimizzazione: SGD, Adagrad, Adadelta
- taglia batch: 32
- numero di epoche: 5

```python
import os
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import SGD, Adagrad, Adadelta

from tensorflow.keras.datasets import mnist
```

#### Caricamento immagini e preparazione dati di input/output

```python
# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits

# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

# reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

# cast floats to single precesion
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255

# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
```

#### Definizione DNN e scelta metodo di ottimizzazione e validazione

```python
def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    
    return model

def compile_model(optimizer):
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=optimizer,
                  metrics=['accuracy'])
    return model

```

#### Training con 3 metodi di ottimizzazione

```python
# metodi di ottimizzazione
Adagrad = keras.optimizers.Adagrad(learning_rate=0.01)
Adadelta = keras.optimizers.Adadelta(learning_rate=1.0, rho=0.95)
SGD = keras.optimizers.SGD(learning_rate=0.01, momentum=0.0, nesterov=False)

opt = [SGD,Adagrad,Adadelta]

# training parameters
batch_size = 32
epochs = 5

# create the deep neural net
model_DNN1 = compile_model(opt[0])
model_DNN2 = compile_model(opt[1])
model_DNN3 = compile_model(opt[2])
```

```python
# model summary 
model_DNN1.summary()
```

```python
# train DNN and store training info in history
history1 = model_DNN1.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
history2 = model_DNN2.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
history3 = model_DNN3.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

#### Validazione modelli

```python
# evaluate model
score1_test = model_DNN1.evaluate(X_test, Y_test, verbose=0)
score2_test = model_DNN2.evaluate(X_test, Y_test, verbose=0)
score3_test = model_DNN3.evaluate(X_test, Y_test, verbose=0)
```

```python
score1_train = model_DNN1.evaluate(X_train, Y_train, verbose=0)
score2_train = model_DNN2.evaluate(X_train, Y_train, verbose=0)
score3_train = model_DNN3.evaluate(X_train, Y_train, verbose=0)
```

#### Training history e confronto performance

```python
# print performance
print('\nSGD MODEL')
print('Train loss:', score1_train[0])
print('Train accuracy:', score1_train[1])
print('Test loss:', score1_test[0])
print('Test accuracy:', score1_test[1])
print('\nADAGRAD MODEL')
print('Train loss:', score2_train[0])
print('Train accuracy:', score2_train[1])
print('Test loss:', score2_test[0])
print('Test accuracy:', score2_test[1])
print('\nADADELTA MODEL')
print('Train loss:', score3_train[0])
print('Train accuracy:', score3_train[1])
print('Test loss:', score3_test[0])
print('Test accuracy:', score3_test[1])
```

```python
history = [history1,history2,history3]

title = ['SGD','Adagrad','Adadelta']
f = plt.figure(figsize = (13,18))

for i in range(3):
    
    f.add_subplot(3,2,2*i+1)
    # summarize history for accuracy
    plt.plot(history[i].history['accuracy'])
    plt.plot(history[i].history['val_accuracy'])
    plt.ylabel('model accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='best')
    plt.title(title[i])
    
    f.add_subplot(3,2,2*i+2)
    # summarize history for loss
    plt.plot(history[i].history['loss'])
    plt.plot(history[i].history['val_loss'])
    plt.ylabel('model loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='best')
    
plt.show()
```

Come ci si aspettava, la funzione di costo diminuisice e l'accuratezza aumenta con i metodi di ottimizzazione Adagrad e Adadelta - mantenendo fissi gli altri parametri. 

Quello che si osserva dai grafici delle training/test histories è che l'accuratezza calcolata durante la fase di test aumenta rispetto a quella calcolata in fase di training. Allo stesso modo - a parità di epoca - la funzione di costo di test è minore rispetto a quella di training. Fa eccezione la rete allenata con Adadelta, in cui dopo circa 2,5 epoche le curve di test e di training si incrociano. 
Se invece si confrontano i valori di loss/accuracy di test e di training ottenuti mediante il metodo "evaluate" si osserva la tendenza opposta (che normalmente ci si aspetta): le performance delle reti sui di dati di training sono migliori di quelle sui dati di validazione. Inoltre, i valori di loss/accuracy ottenuti con il metodo "evalutate" sono minori/maggiori rispetto a quelli ottenuti con il metodo "fit".

Questo comportamento delle reti è dovuto all'utilizzo del metodo Dropout in fase di training. Quando il Dropout è attivo, ad ogni aggiornamento dei parametri una certa percentuale di neuroni - scelta in maniera casuale - viene disattivata. In questo modo, la rete è costretta a trovare vie alternative per ottenere uno stesso risultato e si evita di adattare eccessivamente i parametri ai dati di training. Ciò che si ottiene è quindi una rete capace di fare predizioni migliori su nuovi dati. 
Quando si utilizza il metodo "evaluate", invece, il Dropout viene disattivato e la rete può agire al massimo delle sue potenzialità.
