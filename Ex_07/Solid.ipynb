## Funzione di autocorrelazione e tempo di correlazione

Le funzioni di autocorrelazione temporale di $U/N$ e $P$ sono state calcolate per il range di tempo $[0,200]$ e il loro grafico  riportato di seguito.

```python
# funzione di autocorrelazione
def Correlation(t,t_max,quantity): 
    Sum = np.zeros(5)
    
    for i in range(t_max):
        Sum[0] += quantity[i]
        Sum[1] += quantity[i] * quantity[i]

    for i in range(t_max - t):
        Sum[2] += quantity[i] * quantity[i + t]
        Sum[3] += quantity[i]
        Sum[4] += quantity[i + t]
        
    Sum[0] = (Sum[0]/t_max) * (Sum[0]/t_max)
    Sum[1] = Sum[1]/t_max
    Sum[2] = Sum[2]/(t_max - t)
    Sum[3] = Sum[3]/(t_max - t)
    Sum[4] = Sum[4]/(t_max - t)

    return (Sum[2] - Sum[3]*Sum[4])/(Sum[1] - Sum[0])

# funzione esponenziale per fit
def Exp(x,a):
    return np.exp(- x/a)
```

```python
t_max = 100000
N = 200
t1 = np.arange(N)
Chi_epot = []
Chi_pres = []

# caricamento da file: energia e pressione istantanee vs passi MC
epot_instant = np.loadtxt("Solid/Instant/epot_instant.txt", usecols=(0), unpack = 'true')
pres_instant = np.loadtxt("Solid/Instant/pres_instant.txt", usecols=(0), unpack = 'true')

# calcolo funzioni di autocorrelazione 
for i in range(N):
    Chi_epot.append(Correlation(i,t_max,epot_instant))
    Chi_pres.append(Correlation(i,t_max,pres_instant))
```

```python
# fit funzioni di autocorrelazione con funzione esponenziale
p_opt_epot, p_cov_epot = curve_fit(Exp, t1, Chi_epot)
Fit_Chi_epot = Exp(t1,p_opt_epot[0])
p_opt_pres, p_cov_pres = curve_fit(Exp, t1, Chi_pres)
Fit_Chi_pres = Exp(t1,p_opt_pres[0])

Chi = [Chi_epot,Chi_pres]
Fit_Chi = [Fit_Chi_epot,Fit_Chi_pres]
title = ['energia potenziale','pressione']
```

```python
# plot    
f = plt.figure(figsize=((16,5)))    
for i in range(2):
    f.add_subplot(1,2,i+1)
    plt.plot(t1,Chi[i],linewidth=2,label='dati simulazione')
    plt.plot(t1,Fit_Chi[i], label = "fit con esponenziale $e^{-t/\u03C4_{epot}}$",linewidth=2)
    plt.xlabel('t [passi MC]')
    plt.ylabel(' $\chi$ [unit naturali]')
    plt.title('Funzione di autocorrelazione '+title[i]+' vs tempo',size=14)
    plt.grid(True)
    plt.legend(prop={'size':15})

plt.show()

print("tempo di correlazione di U/N: \u03C4_epot =", round(p_opt_epot[0],1))
print("tempo di correlazione di P: \u03C4_pres =", round(p_opt_pres[0],1))
```

Le funzioni di correlazione presentano, come previsto, un andamento approssimabile ad un decadimento esponenziale e un tempo di correlazione $\tau_c \sim 8$.  In $100$ passi MC le correlazioni di $U/N$ e $P$ sono totalmente estinte.

## Data blocking e incertezza statistica

Si riportano i grafici dell'incertezza statistica $\sigma_L$ dei valori medi $\langle U/N\rangle$ e $\langle P\rangle$ stimata con il metodo del blocking in funzione del numero di passi temporali $L$ in ciascun blocco mantenendo il numero totale di passi costante ($M=10^5$). Sono stati scelti 100 valori di $L$ distribuiti uniformemente nel range $[10,5\cdot 10^3]$ ma nei grafici viene mostrato il range in cui $\sigma_L$ aumenta e raggiunge - all'incirca - il valore limite. I valori di $\sigma_L$ sono stati plottati anche in scala logaritmica.

```python
# funzione errore statistico
def Error(AV,AV2,n):
    return math.sqrt((AV2 - AV**2)/n)

# funzione metodo blocking
def Blocking(quantity):
    M = len(quantity) # total MC steps
    n_punti = 100
    
    L = np.zeros(n_punti).astype(int)
    err = np.zeros(n_punti)

    for i in range(n_punti):
        
        L[i] = (i+1)*10
        N = int(M/L[i]) # number of blocks
        ave1 = 0
        ave2 = 0
        for j in range(N):
            mean = 0
            for k in range(L[i]):
                mean += quantity[j*L[i] + k]/L[i]
            ave1 += mean/N
            ave2 += mean*mean/N
        
        err[i] = Error(ave1,ave2,N)
    return L,err;
```

```python
# calcolo errore statistico per diversi valori di L
L,err_epot = Blocking(epot_instant)
L,err_pres = Blocking(pres_instant)
```

```python
err = [err_epot,err_pres]
title = ['energia potenziale','pressione']

# plot
f = plt.figure(figsize=(13,8))

for i in range(2):
    f.add_subplot(2,2,i+1)
    plt.plot(L,err[i],marker='o',markersize=5,linestyle='None') 
    plt.title('Errore statistico vs taglia blocchi - '+title[i],size=12)
    plt.xlabel('L [passi MC]',size=13)
    plt.ylabel('$\sigma_L$ [unit naturali]',size=13)
    plt.xlim(8,3*10**2)
    plt.grid(True)

for i in range(2):
    f.add_subplot(2,2,i+3)
    plt.plot(L,err[i],marker='o',markersize=5,linestyle='None') 
    plt.xlabel('L [passi MC] - log scale',size=13)
    plt.ylabel('$\sigma_L$ [unit naturali]',size=13)
    plt.xscale('log')
    plt.xlim(8,3*10**2)
    plt.grid(True)
plt.show()
```

I grafici mostrano un incremento di $\sigma_L$ per blocchi di dimensione fino a $100-150$ passi. Aumentando la taglia dei blocchi al di sopra di questo valore, $\sigma_L$ non varia in maniera apprezzabile ed  quindi possibile affermare che essa ha raggiunto il suo valore limite, che corrisponde alla vera deviazione standard dalla media. Il risultato trovato  compatibile con il tempo impiegato dalla funzione di correlazione ad approssimare lo zero.

## Plot dei valori medi per Argon e Krypton

Si riportano i grafici di $\langle U/N\rangle$ e $\langle P\rangle$ e relative incertezze statistiche ottenute con simulazioni di $500$ blocchi da $200$ passi. I risultati relativi all'ultimo blocco sono riportati in unit del SI per l'Argon e il Krypton. 

```python
k_B = 1.38065*10**(-23) # J/K
amu = 1.661*10**(-27) # kg/amu

# ARGON: unit di misura SI
sigma_A = 0.34*10**(-9) # m
epsilon_A = 120*k_B # J
T_A = 120 # K
m_A = 39.948*amu # kg
tau_A = sigma_A*m_A**(1/2)*epsilon_A**(-1/2) # s
p_A = epsilon_A*sigma_A**(-3) # Pa

# KRYPTON: unit SI
sigma_K = 0.364*10**(-9) # m
epsilon_K = 164*k_B # J
T_K = 164 # K
m_K = 83.798*amu # kg
tau_K = sigma_K*m_K**(1/2)*epsilon_K**(-1/2) # s
p_K = epsilon_K*sigma_K**(-3) # Pa
```

```python
# caricamento dati: energia potenziale, pressione, funz di distribuzione radiale
ave_epot,err_ave_epot = np.loadtxt("Solid/Average/epot.txt", usecols=(0,1), unpack = 'true')
ave_pres,err_ave_pres = np.loadtxt("Solid/Average/pres.txt", usecols=(0,1), unpack = 'true')

N = len(ave_epot)
t2 = np.arange(N)

averages = [ave_epot,ave_pres]
err_ave = [err_ave_epot,err_ave_pres]
title = ['Energia potenziale media per particella','Pressione media']
ylabel = ['$U/N$ $[J]$','$P$ $[Pa]$']

# plot
f = plt.figure(figsize=(15,5))

for i in range(2):
    f.add_subplot(1,2,i+1)
    plt.errorbar(t2,averages[i],err_ave[i])
    plt.title(title[i],size=15)
    plt.xlabel('# blocchi',size=12)
    plt.ylabel(ylabel[i],size=12)
    plt.grid()
plt.show()

print('<U/N> (unit ridotte):',round(d(ave_epot[N-1]),3),'+/-',round(d(err_ave_epot[N-1]),3))
print('<U/N> (Argon):',round(d(ave_epot[N-1]*epsilon_A),24),'+/-',round(d(err_ave_epot[N-1]*epsilon_A),24),'J')
print('<U/N> (Krypton):',round(d(ave_epot[N-1]*epsilon_K),24),'+/-',round(d(err_ave_epot[N-1]*epsilon_K),24),'J\n')

print('<P> (unit ridotte):',round(d(ave_pres[N-1]),3),'+/-',round(d(err_ave_pres[N-1]),3))
print('<P> (Argon):',round(d(ave_pres[N-1]*p_A),-5),'+/-',round(d(err_ave_pres[N-1]*p_A),-5),'Pa')
print('<P> (Krypton):',round(d(ave_pres[N-1]*p_K),-5),'+/-',round(d(err_ave_pres[N-1]*p_K),-5),'Pa')
```

##  Funzione di distribuzione radiale

Infine, si mostra il grafico del valore medio della funzione di distribuzione radiale $g(r)$, che fornisce informazioni sulla struttura microscopica del sistema, dove la distanza tra due particelle $r$  riportata in unit del SI per l'Argon e il Krypton. Per ottenere la $g(r)$  stato riempito e opportunamente normalizzato un istogramma di $100$ bin nel range $[0,L_{box}/2]$. 

```python
rho = 1.1
n_part = 108
sigma = [sigma_A,sigma_K]
name = ['Argon','Krypton']

gdr,err_gdr = np.loadtxt("Solid/Average/gave.txt", usecols=(1,2), unpack = 'true')

n_bins = len(gdr)
bin_size = (n_part/rho)**(1.0/3.0)/2.0/n_bins
r = np.arange(n_bins)*bin_size

f = plt.figure(figsize=(15,5))

for i in range(2):
    f.add_subplot(1,2,i+1)
    plt.errorbar(r*sigma[i],gdr,err_gdr)
    plt.title('Funzione di distribuzione radiale - '+name[i],size=15)
    plt.xlabel('r [m]',size=12)
    plt.ylabel('g(r)',size=12)
    plt.grid()
plt.show()
```

La distribuzione  in perfetto accordo con quella ottenuta dalla simulazione di dinamica molecolare del sistema nelle stesse condizioni termodinamiche di temperatura e pressione.
