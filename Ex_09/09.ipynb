# Ex. 09 - Algoritmo Genetico

## Il problema del commesso viaggiatore
In questo esecizio è stato implementato un algoritmo genetico per la risoluzione del problema del commesso viaggiatore, che consiste nel trovare il percorso ad anello più breve che permette di visitare un numero $N$ di città, ciascuna una sola volta. In questo caso, in particolare, il problema è stato risolto per due disposizioni differenti delle città sul piano:

- 30 città disposte in maniera casuale su una circonferenza di raggio unitario
- 30 città disposte in maniera casuale in un quadrato di lato unitario

Se si considerano esclusivamente tragitti in linea retta, la lunghezza $L(X)$ di un percorso $X=[1,\dotsb,N]$ è la seguente:

$$L(X)= \sum_{i=1}^N|\vec x_{i+1}-\vec x_i| $$

dove vale la condizione al contorno periodica $\vec x_{N+1}=\vec x_1$. Equivalentemente, si può considerare la lunghezza $L2(X)=\sum_{i=1}^N|\vec x_{i+1}-\vec x_i|^2 $, cosa che è stata fatta in questo esercizio. Il vincolo di visitare ciascuna città una sola volta si realizza generando sequenze di città a partire da permutazioni della sequenza $[1,\dotsb,N]$.

## Algoritmi genetici
Il problema del commesso viaggiatore è un tipico problema di ottimizzazione combinatoria, in cui la quantità da ottimizzare è la lunghezza del tragitto e la soluzione (sequenza di città che corrisponde al tragitto più breve) appartiene ad un set finito di possibili soluzioni (tutte le possibili permutazioni di una sequenza di città). 
Data la complessità del problema, quest'ultimo si presta bene ad essere risolto con metodi di ottimizzazione metaeuristici come, ad esempio, gli algoritmi genetici.

Gli algoritmi genetici si ispirano al principio di selezione naturale ed evoluzione biologica e consistono in algoritmi che valutano diverse possibili soluzioni di partenza (diversi individui di una popolazione) e che, ricombinandole (riproduzione) ed introducendo elementi di disordine (mutazioni genetiche casuali), producono nuove soluzioni (nuove generazioni di individui) che vengono valutate scegliendo le migliori (selezione ambientale) con lo scopo di convergere verso soluzioni ottimali.

Nel caso in esame un individuo $X$ è stato rappresentato da una sequenza (ID) di numeri interi da $0$ a $29$ (le $30$ città) ottenuta come permutazione $[P_0,\dotsb,P_{29}]$ della sequenza $[0,\dotsb,29]$. La "bontà" (grado di adattamento) di ciascun individuo è stata quantificata dalla quantità $L2(X)$, che prende il nome di funzione di costo o fitness.

Dopo aver creato la mappa delle città ed aver identificato ciascuna di esse con un numero da $0$ a $29$ è stata generata una popolazione di $n_{pop}=900$ individui - attraverso combinazioni di semplici permutazioni tra coppie di città - ed è stato verificato che ciascuno di essi soddisfasse i vincoli richiesti dal problema. La popolazione è stata quindi riordinata in ordine crescente di funzione di costo. Infine, la seguente procedura è stata iterata $n_{iter}=300$ volte, ovvero sono state generate $300$ generazioni:

1. selezione di una coppia di individui (genitori) in funzione della loro fitness. Un individuo $i$ è stato selezionato con l'algoritmo  $i=n_{pop}\cdot (1-r)^p $, dove r è un numero casuale nell'intervallo $[0,1)$ ed è stato scelto, dopo vari tentativi, $p=0.3$.
2. crossover degli ID dei genitori con probabilità $1/2$ per la creazione di 2 nuovi individui. Un nuovo individuo ha la prima metà ID uguale ad un genitore e la seconda metà con le città mancanti che vengono disposte nell'ordine in cui compaiono nell'altro genitore.
3. selezione di un algoritmo di mutazione tra pair permutation, shift, group permutation e inversion, ciascuno con probabilità $p_{mut}=7\%$ (le mutazioni sono mutuamente esclusive). Con probabilità pari al $72\%$, invece, i nuovi individui non subiscono alcuna mutazione.
4. ripetizione dei punti 1,2,3 fino alla creazione di una nuova generazione della stessa dimensione della precedente
5. calcolo della fitness di ciascun individuo, riordino della popolazione in base alla fitness, calcolo della fitness mediata sulla metà migliore della popolazione

Grazie alla presenza di algoritmi di selezione in base alla fitness, di crossover genetico e di mutazioni casuali l'algoritmo genetico combina elementi che favoriscono la convergenza con elementi che permettono di diversificare lo spazio di ricerca. 

Gli andamenti delle fitness sono strettamente legati alla scelta dei parametri (probabilità di mutazione, probabilità di incrocio, algoritmo di selezione). Ad esempio, diminuendo la probabilità di mutazione la convergenza alle soluzioni ottimali avviene in un numero inferiore di generazioni e l'andamento delle fitness è molto più regolare. Al contrario, aumentando la probabilità di mutazione, il tempo di convergenza aumenta a causa del disordine introdotto e la fitness media si attesta su un valore maggiore a causa della maggiore variabilità all'interno della popolazione, ma allo stesso tempo l'algoritmo esplora un numero maggiore di possibili soluzioni.

## Mappa delle città 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot delle città

def CircleUp(x):
    return np.sqrt(1-x*x)
def CircleDown(x):
    return -np.sqrt(1-x*x)

f = plt.figure(figsize = (13,6))

x1,y1 = np.loadtxt('circle.start',usecols = (0,1), delimiter = ' ', unpack = 'true')
x2,y2 = np.loadtxt('square.start',usecols = (0,1), delimiter = ' ', unpack = 'true')

f.add_subplot(1, 2, 1)
x_c = np.linspace(-1,1,400)
plt.plot(x1,y1,'o')
plt.plot(x_c,CircleUp(x_c),color = 'green')
plt.plot(x_c,CircleDown(x_c),color='green')
plt.title('città: disposizione random su circonferenza unitaria', size ='13')
plt.xlabel('x')
plt.ylabel('y')

f.add_subplot(1, 2, 2)
x_s = [0,0,1,1,0]
y_s = [0,1,1,0,0]
plt.plot(x2,y2,'o')
plt.plot(x_s,y_s,'-')
plt.title('città: disposizione random nel quadrato [0,1)x[0,1)', size ='13')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

## Città sulla circonferenza
Innanzitutto si mostrano i risultati relativi alle città disposte sulla circonferenza. Poichè in questo caso la soluzione del problema è nota (il percorso più breve è il poligono inscritto nella circonferenza), è possibile verificare - almeno parzialemente - la correttezza dell'algoritmo genetico.

### Fitness
Di seguito sono riportati due grafici:

- quello a destra rappresenta l'andamento della fitness dell'individuo migliore della popolazione in funzione della generazione
- quello a sinistra rappresenta l'andamento del valor medio della fitness ottenuto sulla metà migliore della popolazione in funzione della generazione

```python
# fitness mediata sulla prima metà della popolazione in funzione della generazione
# fitness del miglior individuo in funzione della generazione

f = plt.figure(figsize = (12,5))
ave_fitness, fitness = np.loadtxt("Circle/fitness.dat", usecols=(0,1), delimiter=' ', unpack='true')
n_gen = np.arange(len(fitness))

f.add_subplot(1, 2, 1)
plt.plot(n_gen,ave_fitness, color = 'orange')
plt.xlabel('generazione')
plt.ylabel('$<L>$')
plt.title('fitness media vs generazione',size = '15')
#plt.xlim(-5,200)
plt.grid()

f.add_subplot(1, 2, 2)
plt.plot(n_gen,fitness)
plt.xlabel('generazione')
plt.ylabel('$L$')
plt.title('fitness miglior individuo vs generazione', size = '15')
#plt.xlim(-5,100)
plt.grid()
plt.show()

best_gen = np.argmin(fitness)

print("Fitness media all'ultima generazione: ",ave_fitness[300])
print("Fitness del percorso migliore : ",fitness[best_gen])
```

La fitness del miglior individuo e la fitness media hanno un andamento - nel complesso - decrescente e tendono infine ad assestarsi attorno ad un valore di equilibrio (in meno di $100$ generazioni la fitness del miglior individuo e in $150$ la fintess media) pari a $2.29$.

### Plot del percorso più breve

```python
# plot del percorso più breve 

ID = np.loadtxt('Circle/CirclePath.'+str(best_gen)+'.final',usecols = (0), delimiter = ' ', unpack = 'true')
ID = np.array(ID, dtype=int)

x = np.zeros(31)
y = np.zeros(31)

for i in range(30):
    x[i] = x1[ID[i]]
    y[i] = y1[ID[i]]
    
x[30] = x1[ID[0]]
y[30] = y1[ID[0]]
    
plt.figure(figsize = (6,6))
plt.plot(x,y,'-',color='lightblue',linewidth=2.5)
plt.plot(x,y,'o',color='blue')
plt.show()
    
```

Come ci si aspettava, il percorso più breve (linee azzurre) che collega le città sulla circonferenza (punti blu) è il poligono ad essa inscritto. 

## Città nel quadrato
Il problema diventa un po' più complesso nel caso in cui le città siano disposte nel quadrato (anche perchè non è possibile verificare che l'algoritmo abbia effettivamente trovato la/le soluzione/i corretta/e). Oltre a richiedere un numero di generazioni maggiori per raggiungere la soluzione ottimale, l'algoritmo converge a soluzioni leggermente diverse a seconda del seme con cui viene inizializzato il generatore di numeri casuali. In ogni caso le diverse soluzioni corrispondono a percorsi di lunghezza simile.

### Fitness
Anche in questo caso si mostrano i grafici di fitness e fitness media in funzione del numero di generazioni.

```python
# fitness mediata sulla prima metà della popolazione in funzione della generazione
# fitness del miglior individuo in funzione della generazione

f = plt.figure(figsize = (12,5))
ave_fitness, fitness = np.loadtxt("Square/fitness.dat", usecols=(0,1), delimiter=' ', unpack='true')
n_gen = np.arange(len(fitness))

f.add_subplot(1, 2, 1)
plt.plot(n_gen,ave_fitness, color = 'orange')
plt.xlabel('generazione')
plt.ylabel('$<L>$')
plt.title('fitness media vs generazione',size = '15')
plt.grid()

f.add_subplot(1, 2, 2)
plt.plot(n_gen,fitness)
plt.xlabel('generazione')
plt.ylabel('$L$')
plt.title('fitness miglior individuo vs generazione', size = '15')
plt.grid()
plt.show()

best_gen = np.argmin(fitness)

print("Fitness media all'ultima generazione: ",ave_fitness[300])
print("Fitness del percorso migliore : ",fitness[best_gen])
```

Anche in questo caso l'andamento è complessivamente decrescente fino al raggiungimento, in circa $200$ generazioni, di un valore di equilibrio pari a $1.06$. Avendo condotto la simulazione per un numero elevato di generazioni, fitness media e fitness dell'individuo migliore coincidono.

### Plot del percorso più breve

```python
# plot del percorso più breve 

ID = np.loadtxt('Square/SquarePath.'+str(best_gen)+'.final',usecols = (0), delimiter = ' ', unpack = 'true')
ID = np.array(ID, dtype=int)

for i in range(30):
    x[i] = x2[ID[i]]
    y[i] = y2[ID[i]]
    
x[30] = x2[ID[0]]
y[30] = y2[ID[0]]

plt.figure(figsize = (6,6))
plt.plot(x,y,'-',color='lightblue',linewidth=2.5)
plt.plot(x,y,'o',color='blue')
plt.show()
```
