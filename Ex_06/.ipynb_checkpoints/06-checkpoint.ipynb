# Ex. 06 - Simulazione del modello di Ising 1D: Metropolis vs Gibbs sampling

## Il modello di Ising 1D
In questo esercizio sono stati applicati algoritmi MC per lo studio del modello di Ising (1D), un modello teorico ampiamente utilizzato in meccanica statistica per lo studio semplificato di sistemi magnetici (o di un qualunque sistema appartenente alla stessa classe di universalità). Si tratta di un modello di spin su reticolo in cui ciascuno spin $\sigma_i$ può assumere solo due valori (up/down, -1/1) e l'interazione tra coppie di spin avviene generalmente tra primi vicini. La dinamica di $N$ spin su reticolo è descritta dalla seguente hamiltoniana $H$:

$$H= -\frac{1}{2}\sum_{i,j n.n.}J_{i,j}\sigma_i\sigma_j - \sum_{i}^Nh_i\sigma_i$$

Il primo termine rappresenta l'interazione tra spin primi vicini (la definzione di primo vicino dipende dalla geometria del reticolo) mentre il secondo rappresenta l'interazione di ciascuno spin con un campo esterno $h_i$. Il segno di $J_{i,j}$ determina la tendenza di due spin ad allinearsi parallelamente ($J_{i,j}>0$, caso ferromagnetico) o antiparallelamente ($J_{i,j}<0$, caso antiferromagnetico) tra loro.

In particolare, nel caso più semplice 1D con $J_{i,j}=J\;\forall \;i,j$ e $h_i=h\;\forall\; i$  e avendo posto le pbc ($\sigma_{N+1}=\sigma_1$), l'hamiltoniana assume la seguente forma:

$$H = -J\sum_{i=1}^N \sigma_i \sigma_{i+1}-\frac{h}{2}\sum_{i=1}^N (\sigma_i + \sigma_{i+1})$$

Per ottenere le proprietà del sistema all'equilibrio termodinamico è necessario calcolare la funzione di partizione $Z$, che dipende dai vincoli a cui è sottoposto il sistema. Per un sistema a temperatura $T$ costante $Z=\sum_{\{\sigma_i\}}e^{-H/T}$ e il peso statistico associato a ciascuno stato $\mu$ è quello di Boltzmann: $p(\mu)=e^{-H/T}/Z$. Dalla funzione di partizione è possibile poi ricavare l'energia libera e i valori di aspettazione di tutte le grandezze funzioni degli stati $\mu$ del sistema.

Il modello di Ising 1D è risolvibile analiticamente e non presenta transizioni  fase paramagnetica/ferromagnetica per $(T\neq 0,h=0)$ nel limite termodinamico ($N \rightarrow \infty$). Ciò significa che la fase ferromagnetica (tutti gli spin allineati) si verifica solo a temperatura nulla o in presenza di un campo esterno. In seguito confronteremo i risultati ottenuti dalle simulazioni con i corrispondenti valori di aspettazione teorici delle seguenti grandezze:
1. energia interna $U(N,T) = \langle H \rangle_{h=0}$
2. capacità termica $C(N,T) = \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ 
3. suscettività $\chi(N,T) = \beta \langle (\sum_{i=1,N}\sigma_i)^2 \rangle_{h=0}$
4. magnetizzazione $M(N,T,h=0.02)= \langle \sum_{i=1,N}\sigma_i \rangle_{h=0.02}$


## Gli algoritmi per il campionamento: Metropolis e Gibbs
È stata simulata una catena di $N=50$ spin, ciascuno dei quali interagente con i 2 spin adiacenti in PBC, per $100$ temperature diverse nel range $[0.5,2]$. 

La distribuzione di Boltzmann è stata campionata con due metodi differenti: l'algoritmo di Metropolis (introdotto nel jupyter dell'eserczio 5) e l'algoritmo di Gibbs. Quest'ultimo permette di capionare una distribuzione di probabilità $p(x_1,..,x_N)$ generando - per $\textit una$ variabile $x_i$ $\textit alla$ $ \textit volta$ - una nuova configurazione $x_i'$ a partire dalla distribuzione di probabilità condizionata $p_{1|N-1}(x_i'|\{x_{j\neq i}\})$, che rappresenta la probabilità di effettuare il passo $x_i \rightarrow x_i'$ mantenendo costanti i valori delle restanti variabili. Quest'ultima è legata alla congiunta $p(x_1,..,x_i',..,x_N)$ e alla marginale $p(\{x_{j\neq i}\})=\sum_{x_{i}}p(x_1,..,x_i,..,x_N)$ (caso di distribuzione discreta) attraverso la seguente relazione:

$$ p_{1|N-1}(x_i'|{x_{j\neq i}}) = \frac{p(x_1,..,x_i',..,x_N)}{p(\{x_{j\neq i}\})}= \frac{p(x_1,..,x_i',..,x_N)}{\sum_{x_{i}}p(x_1,...,x_N)}$$

È immediato verificare che scegliendo questa probabilità condizionata come probabilità di transizione, si realizza una catena di markov in cui il bilancio dettagliato è sempre soddisfatto: ciò equivale, nel caso del Metropolis, ad effettuare mosse accettate con probabilità 1. Il vantaggio del campionamento di Gibbs consiste quindi nell'effettuare mosse sempre accettate campionando in maniera diretta la probabilità condizionata di ogni variabile, in genere più semplice da campionare rispetto alla congiunta. Il campionamento di Gibbs è però facilmente applicabile nel caso in cui ciascuna variabile da campionare possa assumere un numero limitato di valori e nel caso in cui la distribuzione condizionata sia facilmente campionabile. 
Il modello di Ising studiato in questo esercizio ($50$ variabili definite sull'insieme $\{-1,1\}$) soddisfa tali richieste. In questo caso, le probabilità condizionate da calcolare per ciascuno spin $\sigma_i$ sono due, $p_{1|N-1}(\sigma_i=\pm1)$:
$$p_{1|N-1}(\sigma_i=\pm1) = \frac{p(\sigma_i=\pm1)}{p(\sigma_i=+1)+p(\sigma_i=-1)}$$

Per entrambi i campionamenti (Metropolis e Gibbs) un singolo passo MC è stato realizzato tentando una mossa per ciascuno spin (quindi $N$ mosse per passo), in maniera tale da limitare le correlazioni tra configurazioni successive. Come al solito, è stato implementato il metodo del blocking per il calcolo delle incertezze statistiche. Per simulare $M=5\cdot10^5$ passi MC sono stati usati $\mathcal{N}=50$ blocchi da $10^4$ passi, ampiamente sufficienti per ottenere blocchi scorrelati tra loro.

## Equilibrazione

Prima di calcolare i valori medi delle varie grandezze è stato necessario assicurarsi che il sistema fosse all'equilibrio, ovvero che si stesse effettivamente campionando la distribuzione di Boltzmann. Ci si è quindi domandati quale fosse il tempo di equilibrazione. Il grafico sottostante mostra l'andamento dell'energia istantanea a $T=1$ in funzione del numero di passi MC (Metropolis) per tre diversi configurazioni di partenza degli spin: 
- configurazione a $T=0$ (tutti gli spin allineati)
- configurazione a $T=\infty$ (spin distribuiti in maniera casuale)
- configurazione a $T=1$, la temperatura di campionamento

Si può osservare che in non più di $10$ passi MC il sistema è all'equilibrio, per qualsiasi configurazione di partenza. L'ideale, comunque, è partire da cofigurazioni prossime a quelle di equilibrio. Per questo motivo si è deciso cominciare con la simulazione a temperatura più alta partendo dalla configurazione disordinata ($T=\infty$) e di proseguire con le simulazioni a temperatura più bassa partendo ogni volta dalla configurazione finale della simulazione precedente. In questo modo non è stato necessario equilibrare il sistema.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
inst_ene = []
inst_ene.append(np.loadtxt("Metropolis/Equilibrazione/instant_ene0.txt", usecols=(0), unpack = 'true'))
inst_ene.append(np.loadtxt("Metropolis/Equilibrazione/instant_ene_inf.txt", usecols=(0), unpack = 'true'))
inst_ene.append(np.loadtxt("Metropolis/Equilibrazione/instant_ene.txt", usecols=(0), unpack = 'true'))
N = len(inst_ene[0])
x = np.arange(N)

# parametri
T = 1.0
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th

# valore teorico energia interna
teo_ene = -J*( th + ch*thN )/( 1 + thN )

label = ['config. iniziale $T=0$','config. iniziale $T=inf $','config. iniziale $T=1$']
color = ['lightgreen','orange','teal']

plt.figure(figsize=(8,5))
# plot energia istantanea vs passi MC a T=1 per 3 configurazioni di partenza diverse: T=0,T=inf,T=1
for i in range(3):
    plt.plot(x,inst_ene[i]/Ns,label=label[i],linewidth=2,color=color[i])
    plt.plot(0,inst_ene[i][0]/Ns,marker='*',markersize=15,color=color[i])
plt.plot(x,np.ones(N)*teo_ene,label='valore di aspettazione teorico',linestyle='dashed',linewidth=2,color='red')
plt.xlabel('MC steps')
plt.xlim(-0.5,50)
plt.ylabel(' E ')
plt.title('Energia istantanea a $T=1$ vs passi MC per diverse configurazioni di partenza',size=14)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# valori di aspettazione teorici di energia interna (h=0), calore specifico (h=0), suscettività (h=0)
# e magnetizzazione (h=0.02) in funzione della temperatura nel range [0.2,2.2]

points = 100
T = np.linspace(2.2,0.2,num=points)

# Calcolo valori medi teorici
b = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
h=0.02

# energia interna
energy = -J*( th + ch*thN )/( 1 + thN )
# calore specifico
heat = ((b*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
# suscettività
chi = b*np.exp(2*b*J)*(1-thN)/(1+thN)
# magnetizzazione
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
magne = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

```

## Risultati

Di seguito sono mostrati i grafici dei valori medi di energia interna per spin, calore specifico, suscettività e magnetizzazione per spin in funzione della temperatura ottenuti mediante i due metodi di campionamento, Gibbs e Metropolis. Sono state plottate anche le curve teoriche per verificare la correttezza dei risultati.

```python
f = plt.figure(figsize=(15,11))

Gibbs_data = []
Metro_data = []
Theo_data = [energy,heat,chi,magne]

t1 = np.loadtxt("Gibbs/Energy.txt", usecols=(0), unpack = 'true')

# Caricamento valori medi ottenuti con Gibbs
Gibbs_data.append(np.loadtxt("Gibbs/Energy.txt", usecols=(1,2), unpack = 'true'))
Gibbs_data.append(np.loadtxt("Gibbs/Heat.txt", usecols=(1,2), unpack = 'true'))
Gibbs_data.append(np.loadtxt("Gibbs/Chi.txt", usecols=(1,2), unpack = 'true'))
Gibbs_data.append(np.loadtxt("Gibbs/Mag.txt", usecols=(1,2), unpack = 'true'))

# Caricamento valori medi ottenuti con Metropolis
Metro_data.append(np.loadtxt("Metropolis/Energy.txt", usecols=(1,2), unpack = 'true'))
Metro_data.append(np.loadtxt("Metropolis/Heat.txt", usecols=(1,2), unpack = 'true'))
Metro_data.append(np.loadtxt("Metropolis/Chi.txt", usecols=(1,2), unpack = 'true'))
Metro_data.append(np.loadtxt("Metropolis/Mag.txt", usecols=(1,2), unpack = 'true'))


title = ['Energia interna $U$ ($h=0$)','Calore specifico $C$ ($h=0$)','Suscettività magnetica $\chi$ ($h=0$)','Magnetizzazione $M$ ($h=0.02$)']
ylabel = ['$U$','$H$','$\chi$','$M$']

# plot dei valori medi di energia interna, calore specifico, suscettività e magnetizzazione in funzione della temperatura
# ottenuti con algoritmo Metropolis e Gibbs 
# confronto con soluzioni analitiche
for i in range(4):
    f.add_subplot(2,2,i+1)
    plt.errorbar(t1,Gibbs_data[i][0],yerr=Gibbs_data[i][1],label = 'Gibbs',color='green')
    plt.errorbar(t1,Metro_data[i][0],yerr=Metro_data[i][1],label = 'Metropolis',color='red')
    plt.plot(T,Theo_data[i],label = 'Andamento teorico', linewidth=2,color='darkorange')
    plt.xlabel('temperatura T')
    plt.ylabel(ylabel[i])
    plt.title(title[i],size=15)
    plt.legend()
    plt.grid(True)

plt.show()
```

Ciò che si evince dai grafici, specialmente da quelli della suscettività e della magnetizzazione, è che a basse temperature ($T<1$) i valori simulati si discostano maggiormente da quelli teorici e gli errori sono più grandi. Ciò significa che il campionamento è più difficoltoso e in effetti il processo stocastico che dovrebbe condurre al campionamento della probabilità di Boltzmann è scarsamente ergodico a basse temperature.
