# Ex. 5 - Campionamento di orbitali atomici con l'algoritmo Metropoli

## L'algoritmo Metropolis
Data una distribuzione di probabilità $p(x)$ da campionare, dove $x$ è un vettore di dimensione arbitraria, l'algoritmo Metropolis consiste nel realizzare un random walk (più in generale un processo stocastico markoviano stazionario) nello spazio delle $x$, in modo tale che, passo dopo passo, la distribuzione di probabilità asintotica di $x$ sia $p(x)$. È possibile dimostrare che, per soddisfare tale richiesta, la probabilità di transizione $K(x'|x)$ (detta nucleo stocastico) che definisce il processo stocastico di Markov stazionario deve soddisfare alcune proprietà. In particolare, è sufficiente che $K(x'|x)$ sia regolare e soddisfi il bilancio dettagliato ($K(x'|x)p(x)=K(x|x')p(x')$). Per soddisfare tale richiesta, l'algoritmo Metropolis decompone la probabilità di transizione $K(x'|x)$ nel prodotto di una probabilità di transizione di prova $T(x'|x)$ e di una probabilità $A(x'|x)$ di accettazione della mossa di prova $x \rightarrow x'$:

$$K(x'|x)=T(x'|x)\times A(x'|x)$$

Se viene scelta una probabilità di prova simmetrica, tale cioè che $T(x'|x)=T(x |x')$, il seguente nucleo stocastico $K(x'| x)$ converge alla distribuzione $p(x)$ da campionare:

$$A(x'|x)=min\big[1,\frac{p(x')}{p(x)}\big]$$

Due caratteristiche rendono l'algoritmo Metropolis particolarmente potente:

- potenzialmente permette di campionare una $\textit qualsiasi$ distribuzione di probabilità $p(x)$
- non necessita della conoscenza della $p(x)$ normalizzata, dato che la probabilità di accettazione di una mossa $x \rightarrow x'$ dipende solo dal $\textit rapporto$ $\frac{p(x')}{p(x)}$.

Ci sono comunque alcuni svantaggi legati all'algoritmo Metropolis:

- il processo stocastico alla base del campionamento converge solo asintoticamente alla distribuzione desiderata. Ciò rende necessaria una fase di equilibrazione e, più in generale, mosse che permettano di esplorare in maniera davvero ergodica lo spazio da campionare.
- configurazioni generate in successione sono generalmente fortemente correlate tra loro. Per la stima del valore di aspettazione di qualsiasi osservabile è opportuno utilizzare il metodo del blocking e scegliere la taglia di ciascun blocco in base al "tempo" di decadimento delle correlazioni.

Affinchè l'algoritmo converga alla distribuzione desiderata (e lo faccia in modo efficiente) è quindi importante scegliere attentamente il tipo di mossa da compiere (ovvero la $T(x'|x)$). In generale, se la dimensione del passo è molto più grande della scala su cui varia la distribuzione $p(x)$ le mosse $x \rightarrow x'$ saranno spesso rigettate ($\frac{p(x')}{p(x)}\ll1$) e, di conseguenza, il campionamento sarà molto lento e inefficiente. Al contrario, se la dimensione del passo è troppo piccola l'accettazione sarà molto alta ($\frac{p(x')}{p(x)}\simeq1$) e le configurazioni molto correlate tra loro. Come regola empirica, un'accettazione del 50% permette di ottenere risultati soddisfacenti.

## Il problema

In questo esercizio è stato utilizzato l'algoritmo Metropolis per campionare la distribuzione di probabilità della posizione dell'elettrone nell'atomo di idrogeno a partire dalla funzione d'onda $\Psi(\vec{r},t)$ che ne rappresenta lo stato. In particolare, le funzioni d'onda rappresentative degli autostati dell'energia dell'elettrone dell'atomo di idrogeno sono le seguenti:

$$
\Psi_{n,l,m}(r,\theta,\phi)=
\sqrt{\left(\frac{2}{na_0}\right)^3 \frac{(n-l-1)!}{2n[(n+l)!]}}
e^{-r/na_0}\left(\frac{2r}{na_0}\right)^l L_{n-l-1}^{2l+1}\left(\frac{2r}{na_0}\right)
Y_l^m(\theta,\phi)
$$

dove $(n,l,m)$ sono i numeri quantici che identificano univocamente lo stato, $(r,\theta,\phi)$ è la posizione dell'elettrone in coordiante sferiche, le funzioni $L_{n-l-1}^{2l+1}\left(\frac{2r}{na_0}\right)$ e $Y_l^m(\theta,\phi)$ identificano rispettivamente i polinomi di Lagerre e le armoniche sferiche, $a_0=4\pi\epsilon_0\hbar^2/m_e e^2=0.0529$ nm è il raggio di Bohr. 
Il modulo quadro $|\Psi_{n,l,m}(r,\theta,\phi)|^2$  della funzione d'onda rappresenta la densità di probabilità della posizione dell'elettrone.

In particolare, sono stati campionati l'orbitale di ground state $1s$ ($n=1$, $l=0$, $m=0$) e uno degli orbitali $2p$ ($n=2$, $l=1$, $m=0$) rappresentati dalle seguenti funzioni d'onda:

$\Psi_{1,0,0}(r,\theta,\phi)=\frac{a_0^{3/2}}{\sqrt{\pi}} e^{-a_0 r} $

$\Psi_{2,1,0}(r,\theta,\phi)=\frac{a_0^{5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-a_0 r/2} \cos(\theta)$

Da cui derivano le seguenti distribuzioni:

$p_{1s}(r,\theta,\phi)=|\Psi_{1,0,0}(r,\theta,\phi)|^2 \propto e^{-2a_0r}$

$p_{2p}(r,\theta,\phi)=|\Psi_{2,1,0}(r,\theta,\phi)|^2 \propto r^2e^{-a_0r}cos^2(\theta)$


## Campionamento con mosse uniformi
Le distribuzioni $p_{1s}$ e $p_{2p}$ sono state campionate in coordinate cartesiane utilizzando una probabilità di transizione $T(x'|x)$ uniforme, ovvero un random walk 3D nel continuo caratterizzato da passi di lunghezza fissata. A seconda della distribuzione da campionare, il passo è stato scelto - per tentativi - in maniera tale da ottenere un'accettazione prossima al 50%. Per definire l'ordine di grandezza del passo è stato utilizzato il raggio di Bohr $a_0$, che definisce la scala di lunghezza tipica del sistema. In particolare, tutte le lunghezze sono state espresse in unità di $a_0$ per ottenere valori di ordine $\sim 1$. La dimensione del passo è stata scelta pari a $1.1$ per il campionamento di $p_{1s}$ e pari a $2.5$ per il campionamento di $p_{2p}$.

Le simulazioni sono state realizzate con $M=10^5$ passi MC suddivisi in $N=10^2$ blocchi da $10^3$ passi ciascuno ed è stato calcolato il valore medio della distanza dall'origine $\langle r \rangle$ in funzione del numero di blocchi. I risultati sono stati poi confrontati con i rispettivi valori di aspettazione teorici, che sono:

- $\langle r \rangle_{1s}=1.5$
- $\langle r \rangle_{2p}=5$

L'errore statistico è stato stimato con l'usuale metodo del blocking. Come mostrato di seguito, la taglia dei blocchi eccede di molto il tempo di decadimento delle correlazioni. 

Di seguito sono mostrati alcuni grafici relativi all'equilibrazione del sistema e - per ciascuna simulazione:

- lo scatterplot di $10^4$ punti ottenuti dal campionamento della distribuzione $p$
- il grafico della stima della distanza media dall'origine $\langle r \rangle$ - con relativa incertezza statistica - in funzione del numero di blocchi

### L'equilibrazione
Ad esclusione di pochi casi, nel realizzare una simulazione Monte Carlo è importante preoccuparsi dell'equilibrazione del sistema, ovvero assicurarsi che le configurazioni generate siano distribuite effettivamente secondo la distribuzione di probabilità desiderata. In particolare, l'algoritmo Metropolis, grazie al suo metodo di accettazione/rigetto della mossa, permette di generare un processo stocastico che solo $\textit asintoticamente$ converge a quello desiderato. Il numero di passi necessario ad equilibrare il sistema dipende dalla distribuzione che si vuole campionare, dalla configurazione di partenza e dal tipo di mossa. 

In tutti i casi presentati in questo esercizio è stato sufficiente partire dalla configurazione $(0,0,1)$ per riuscire ad equilibrare il sistema in pochi passi. Chiaramente, il punto di partenza va scelto in base alla distribuzione da campionare. 

Di seguito sono mostrati due grafici della distanza istantanea dall'origine in funzione del numero di step MC a partire da configurazioni diverse: $(0,0,1)$ a destra e $(10,10,10)$ a sinistra. Entrambi si riferiscono al campionamento dell'orbitale 1S con passi uniformi di lunghezza $1.1$. Il grafico a sinistra mostra in maniera chiara come una configurazione di partenza "poco probabile" richieda un tempo di equilibrazione più lungo (circa $100$ passi), rispetto ad una configurazione probabile.

I grafici mostrano anche che, adottando un passo di dimensione pari a $1.1$, che corrisponde ad un'accettazione $\sim$ 50%, le correlazioni decadono in pochi passi. Questo rapido decadimento delle correlazioni si è verificato in tutte le simulazioni effettuate, avendo avuto cura di adattare la dimensione del passo ad ogni caso.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import math
import decimal
```

```python
# 1S passi uniformi: plot della distanza istantanea dall'origine (equilibrazione)
f = plt.figure(figsize=(14,4))

f.add_subplot(1,2,1)
r = np.loadtxt("Uniform/1S_instant_r1.txt", usecols=(0), delimiter = ' ', unpack = 'true')
N = len(r)
x = np.arange(N)

plt.plot(x,r)
plt.plot(x,np.ones(N)*1.5,c='r',label='$<r>_{1s}$ atteso')
plt.xlabel('step')
plt.ylabel(' r ')
plt.title('distanza istantanea dall\'origine vs passo MC - partenza in (10,10,10)')
plt.legend()
plt.grid(True)

f.add_subplot(1,2,2)
r = np.loadtxt("Uniform/1S_instant_r2.txt", usecols=(0), delimiter = ' ', unpack = 'true')
N = len(r)
x = np.arange(N)

plt.plot(x,r)
plt.plot(x,np.ones(N)*1.5,c='r',label='$<r>_{1s}$ atteso')
plt.xlabel('step')
plt.ylabel(' r ')
plt.title('distanza istantanea dall\'origine vs passo MC - partenza in (0,0,1)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 1S passi uniformi : scatterplot dei punti campionati
X,Y,Z = np.loadtxt("Uniform/1S_xyz.txt", usecols=(0,1,2), delimiter = ' ', unpack = 'true',max_rows = 10**4)
d = len(X)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111,projection='3d')

#ax = Axes3D(fig)
ax.scatter(X, Y, Z, s=5,c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.title('Campionamento $|\Psi_{1,0,0}(x,y,z)|^2$',size=17)
ax.view_init(10, 30)
plt.show()
```

## Campionamento con mosse gaussiane

Le simulazioni sono state ripetute utilizzando una probabilità di transizione $T(x'|x)$ di tipo gaussiano invece di una probabilità uniforme. Ad ogni istante, quindi, per ciascuna coordinata è stato generato uno step distribuito come una gaussiana centrata in $0$. Per ottenere un campionamento simile a quello ottenuto in precedenza (con 50% di accettazione), è stato necessario scegliere accuratamente la varianza $\sigma$, che, per $\textit ogni$ $\textit coordinata$, rappresenta la lunghezza media di uno step $\omega$ (cioè la distanza media tra due configurazioni successive). In 3 dimensioni la distanza media tra due configurazioni successive vale quindi:

$$\big\langle \big(\vec x_{i+1}-\vec x_{i}\big)^2\big\rangle = \big\langle \sum_{\alpha}^3\omega_{\alpha}^2 \big\rangle = 3\sigma^2$$

Quindi, se per il campionamento uniforme si è scelto $\big\langle \big(\vec x_{i+1}-\vec x_{i}\big)^2\big\rangle = a^2$, nel caso del campionamento gaussiano bisogna scegliere $\sigma\sim\frac{a}{\sqrt3}$.

Questa regola è stata effettivamente verificata durante la simulazione, in cui sono stati utilizzati valori di $\sigma$ pari a $0.8$ e $1.8$ per il campionamento rispettivamente di $p_{1s}$ e $p_{2p}$.

Di seguito sono mostrati - per ciascuna simulazione:

- lo scatterplot di $10^4$ punti ottenuti dal campionamento della distribuzione $p$
- il grafico della stima della distanza media dall'origine $\langle r \rangle$ - con relativa incertezza statistica - in funzione del numero di blocchi
