```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.optimize import curve_fit

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])
print("1st excited state energy: ", E[1])
print("2nd excited state energy: ", E[2])
```

```python

# ANALITIC CURVE OF PROBABILITY DISTRIBUTION (SUM OF 2 GAUSSIANS with OPTIMIZED PARAMETERS mean = 0.8 sigma = 0.62) 
# COMPARISON WITH PROBABILITY DISTRIBUTION OBTAINED WITH THE CENTRAL DIFFERENCES METHOD

mean = 0.8
sigma = 0.62

def Psi2(x):
    const = 2 * sigma * np.sqrt(np.pi) * ( 1 + np.exp( - mean**2 / sigma**2 ) )
    return ( np.exp( - ( x - mean )**2 / sigma**2 ) + np.exp( - ( x + mean )**2 / sigma**2 ) + 2 * np.exp( - ( x**2 + mean**2) / sigma**2 ) )/const

PSI2 = Psi2(x)

# Plot a few things
plt.figure(figsize=(8,5))
scale = 0.3
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.plot(x,PSI2, color="Red", label="Psi2 - optimized analitic curve")
plt.plot(x,(psi[0])**2, label = "Psi2 - central diff. method")
#plt.plot(x,(psi[1])**2)
#plt.plot(x,(psi[2])**2)
plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.legend()
plt.show()
```

### PIGS - PATH INTEGRAL GROUND STATE ($T = 0 K$)


Singola particella in potenziale 1D: $V(x) = x^4 - \frac{5}{2} x^2$.


Funzione d'onda di prova: $\psi_0(x) = 1$
