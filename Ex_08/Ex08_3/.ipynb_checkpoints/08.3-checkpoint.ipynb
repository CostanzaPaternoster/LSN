## Path Integral Ground State (PIGS) e Path Integral Monte Carlo (PIMC)

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.optimize import curve_fit
```

```python
# distribuzione di probabilità doppia gaussiana
def DoubleGauss(x, a, b):
    return 1/(b**2 * math.pi * (2 + np.exp(-a**2/b**2)))*(np.exp(-0.5*(x-a)**2/b**2) + 
                                                          np.exp(-0.5*(x+a)**2/b**2))**2
# potenziale 
def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 10
N = 1000

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
```

## PIGS - Calcolo "esatto" dell'energia di ground state (T = 0)


Nell'esercizio precedente si è fatto uso dei metodi variazionali per approssimare il ground state del sistema composto da una particella quantistica in potenziale $V(x)=x^4 - \frac{5}{2} x^2$. La funzione di prova $\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+ e^{-\frac{(x+\mu)^2}{2\sigma^2}}$ è stata scelta in base all'andamento del potenziale e, per quanto buona possa essere la scelta dei parametri $\mu$ e $\sigma$, $\Psi_T^{\sigma,\mu}(x)$ rimane una soluzione approssimata. 

Nonostante non sia possibile ottenere l'esatta funzione d'onda di ground state, esiste una classe di metodi Monte Carlo - noti come Path Integral Projector Methods (PIPM) - che permette di stimare il valore di aspettazione dell'energia sul ground-state (ovvero a temperatura $T = 0 K$) con un errore molto più piccolo dell'errore statistico associato all'uso di metodi MC (per questo motivo si parla di calcolo "esatto"). 

Il PIGS, che rientra nella classe dei PIPM, permette di mappare il sistema quantistico a temperatura nulla in un equivalente sistema classico a dimensionalità più alta, in cui le medie sul ground state diventano medie canoniche. 

Per ottenere questo risultato, il PIGS sfrutta l'evoluzione a tempi immaginari di un generico stato quantistico descritta dall'equazione di Schrödinger - che diventa in quest'ottica del tutto equivalente ad un'equazione di diffusione. Inoltre, è possibile mostrare come l'evoluzione di uno stato per tempi immaginari sufficientemente grandi sia proporzionale al ground state del sistema, a noi ignoto.

Indicando con $|\psi_0\rangle$ lo stato del sistema al tempo $t=0$ (che chiameremo stato di prova), con $H$ l'Hamiltoniana del sistema, con $\{ |E_\alpha\rangle \}_{\alpha=0}^{\infty}$ una base di autostati di $H$ e con $E_\alpha$ i corrispondenti autovalori, si ha:

$$
|\psi(\tau) \rangle := e^{-\hat H \tau} |\psi_0\rangle =  e^{-\hat H \tau} \sum_{\alpha = 0}^\infty | E_\alpha \rangle \langle E_\alpha |\psi_0 \rangle  = \sum_{\alpha = 0}^\infty c_\alpha e^{- E_\alpha \tau}  |E_\alpha\rangle \approx_{\tau\gg1}c_{0}e^{-\tau E_{0}}|GS\rangle
\tag{1}
$$ 
$\text{con } c_\alpha = \langle E_\alpha|\psi_0\rangle$

dove $\tau$, il tempo immaginario, si ottiene mediante la cosiddetta rotazione di Wick ($t\rightarrow\frac{it}{h}=\tau$).

Una qualsiasi funzione d'onda di prova può quindi essere mappata - in linea teorica - sul ground state del sistema. Il tempo immaginario necessario per ottenere la proiezione dipenderà dai coefficienti $c_{\alpha}$, ovvero dal grado di sovrapposizione dello stato di prova con gli autostati di $H$. In questo modo è anche possibile riscrivere il valore di aspettazione di una osservabile $\hat O$ sul ground state come:

$$
\langle GS | \hat O | GS \rangle = \underset{\tau \rightarrow \infty}{\lim}\frac{\langle \psi_0 | e^{-\hat H \tau }\hat O e^{-\hat H \tau }| \psi_0 \rangle}{\langle \psi (\tau)|\psi (\tau) \rangle}
\tag{2}
$$

Attraverso la formulazione path integral è possibile poi esprimere il propagatore $G(x,x',t)=\langle x|e^{-\tau H}|x'\rangle$ - che compare se rappresentiamo l'evoluzione dello stato sulla base delle coordinate - in funzione di quantità note. Questo permette di valutare il valore di aspettazione $(2)$. Nel caso in cui $\hat O$ sia diagonale sulla base delle coordinate, il risultato che si ottiene, per 1 particella in 1 dimensione, è il seguente:

$$
\langle GS| \hat O |GS \rangle \approx \int \Big(\prod\limits_{i=1}^{2M+1} dx_i \Big) O(\cal x_{\textrm{M+1}}) \hspace{0.1 cm} \underbrace{\frac{\psi_0 (x_1) \hspace{0.1 cm}\psi_0 (x_{\textrm{2M+1}})}{\langle \psi (\tau)|\psi (\tau) \rangle} \hspace{0.1 cm} \prod\limits_{i=1}^{\textrm{2M}} G(x_i,x_{i+1}, \frac{\tau }{ \textrm M})}_{p(x_{1},...,x_{2M+1})} \tag{3}
$$

dove

$$
G(x_{i}, x_{i+1}, \delta \tau) = \langle x_{i}|e^{-\delta\tau(\hat T+\hat V)}|x_{i+1}\rangle = e^{-\frac{\delta \tau}{2} V( x_{i})}\frac{exp\big({\frac{| x_{i}- x_{i+1}|^2}{4\delta \tau\gamma}}\big)}{(4\pi\delta \tau\gamma)^{3/2}}e^{-\frac{\delta \tau}{2} V( x_{i+1})} \tag{4} 
$$ 

$\delta\tau = \frac{\tau}{M}, \hspace{0.2 cm} \gamma=\hbar/2m$

$M$ è un parametro che compare nella procedura di path integral e che può essere interpretato come il numero di intervalli in cui viene diviso il tempo immaginario $\tau$. All'aumentare di $M$, il termine di dx della $(3)$ approssima sempre meglio il termine di sx, ma il numero di integrazioni aumenta come $2M+1$. 

È importante tenere conto delle approssimazioni fatte per ottenere questo risultato:

- il $\lim{\tau \rightarrow \infty}$ viene sostituito con un tempo immaginario $\tau$ finito
- i propagatori nella $(4)$ - che si ottengono a partire dalla decomposizione di Trotter $(e^{-\frac{\tau}{M}\hat H})^{M} = \underset{M\rightarrow \infty}{\lim}(e^{-\frac{\tau}{M}\hat T}e^{-\frac{\tau}{M}\hat V})^{M}$ - vengono valutati separando parte cinetica da parte potenziale grazie all'approssimazione primitiva $\langle x_{i}|e^{-\frac{\tau}{M}(\hat T+\hat V)}|x_{i+1}\rangle \approx \langle x_{i}|e^{-\frac{\tau}{M}\hat V/2}e^{-\frac{\tau}{M}\hat T}e^{-\frac{\tau}{M}\hat V/2}|x_{i+1}\rangle$. Poichè l'errore che si compie è pari a $\frac{1}{2}(\frac{\tau}{M})^2[\hat T,\hat V]$, l'approssimazione è valida se $\delta \tau = \frac{\tau}{M} \ll 1$, ovvero se si sceglie un valore di $M$ sufficiente grande. 

Poichè la quantità che nella $(3)$ moltiplica l'osservabile $O(x_{M+1})$ è normalizzata, il valore di aspettazione $(3)$ è equivalente ad una media classica con distribuzione di probabilità $p(x_{1},...,x_{2M+1}) = \frac{\psi_0 (x_1) \hspace{0.1 cm}\psi_0 (x_{\textrm{2M+1}})}{\langle \psi (\tau)|\psi (\tau) \rangle} \hspace{0.1 cm} \prod\limits_{i=1}^{\textrm{2M}} G(x_i,x_{i+1}, \frac{\tau }{ \textrm M})$, che può quindi essere campionata con un algoritmo Monte Carlo, come il Metropolis. Poichè le variabili di integrazione sono passate da $1$ a $2M+1$, questo significa campionare la posizione di $2M+1$ particelle classiche. In particolare, queste interagiscono con interazione di tipo elastico (di origine cinetica) in modo da formare uno speciale polimero.


### Parametri
Sono state condotte tre simulazioni MC della lunghezza di $20$ blocchi da $4000$ passi ciascuno a partire da tre funzioni d'onda diverse: 
* $\psi_0(x) = 1$
* $\psi_0(x) = e^{-\frac{x^2}{2}}$
* $
\Psi_0^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$ con parametri $\mu = 0.8$ e $\sigma = 0.61$ ottenuti ottimizzando la funzione d'onda di prova nell'esercizio di VMC

La parte cinetica della distribuzione di probabilità - che è una gaussiana - è stata campionata in maniera diretta, mentre per la parte relativa al potenziale si è fatto uso dell'algoritmo Metropolis e diverse possibili "mosse", i cui parametri sono stati scelti in modo da ottenere rate di accettazione $\sim 50\%$.

* numero di BrownianBridge reconstructions $BB = 110$
* numero di BrownianMotion reconstructions $BM = 65$
* passo traslazioni $\delta = 0.55$

Sono stati adottati i seguenti parametri per la proiezione delle funzioni d'onda sul gs:

* tempo immaginario $\tau = 8$
* numero di intervalli di tempo immaginario (numero di particelle del polimero) $M = 300$

Un tempo immaginario $\tau$ pari a 8  e un numero di posizioni del polimero M pari a 300 sono risultati sufficienti per ottenere la proiezione sul GS in tutti e i tre casi. Aumentare il tempo immaginario permette di ottenere una proiezione migliore sul GS e - a seconda dell'overlap tra funzione d'onda di partenza e GS - questo non può scendere al di sotto di un certo valore, indipendentemente dal valore di M. Un numero abbastanza grande di intervalli temporali M è invece necessario per ottenere una buona approssimazione del propagatore a partire dalla decomposizione di Trotter. In particolare, deve essere $(\frac{\tau}{M})^2 << 1$.

### Campionamento del modulo quadro della funzione d'onda di ground state
Di seguito si riportano gli istogrammi delle distribuzioni di probabilità della posizione relative alla funzione di ground state ottenuta con il metodo PIGS per le tre diverse funzioni d'onda di partenza. Queste si confrontano con la distribuzione ottenuta con il metodo di discretizzazione e con la curva ottenuta dal fit con una somma di gaussiane (usata come funzione d'onda di prova nell'esercizio di VMC). Gli istogrammi sono stati realizzati utilizzando le posizioni centrali del polimero.

```python
# plot istogramma distribuzione di probabilità campionata
# funzioni d'onda di prova: costante, gaussiana, doppia gaussiana VMC
# plot distribuzione ottenuta con metodo di discretizzazione

plt.figure(figsize=(10,6))
filename = ['const','doublegauss','gauss']
f_start = ['$\\propto 1$','$=\\psi^{\\sigma,\\mu}(x)$','$\\propto e^{-\\frac{x^2}{2}}$']

for i in range(3):
    bins,histo,err_histo = np.loadtxt("PIGS/probability_"+filename[i]+".dat",usecols = (0,1,2), unpack = 'true')
    plt.errorbar(bins, histo, err_histo, label = "$\psi_0(x)$"+f_start[i]+" ")

    #distribuzione ottenuta con metodo numerico
plt.plot(x,(psi[0])**2, label = "metodo numerico", c= 'black')

#fit con somma di gaussiane 
p_opt, p_cov = curve_fit(DoubleGauss, bins, histo)
y_fit = DoubleGauss(bins,p_opt[0],p_opt[1])
plt.plot(bins,y_fit, label='fit somma gaussiane') 

plt.xlabel('Posizione [unità naturali]', size = 12)
plt.xlim(-2.5, 2.5)
plt.title('Densità di probabilità', size = 18)
plt.grid()
plt.legend()
plt.show()

print("Parametri ottimali del fit con doppia gaussiana:")
print("\u03BC = ", round(p_opt[0],2))
print("\u03C3 = ", round(p_opt[1],2))
print("Parametri ottenuti con metodi variazionali:")
print("\u03BC = ", 0.80)
print("\u03C3 = ", 0.61)
```

Si può notare come la somma di due gaussiane $\Psi^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+e^{-\frac{(x+\mu)^2}{2\sigma^2}}$ - funzione di prova utilizzata nel Variational Monte Carlo per stimare il GS - sia un'approssimazione più grossolana rispetto alla funzione d'onda ottenuta con il metodo "esatto" PIGS. Infatti, le distribuzioni campionate utilizzando il metodo PIGS convergono in entrambi i casi (funzione di partenza costante e doppia gaussiana) ad uno stesso risultato - che si ottiene anche con il metodo di discretizzazione - entro l'errore. Si nota inoltre che gli errori ottenuti a partire da funzioni d'onda di prova diverse non differiscono tra loro in maniera significativa. Ciò significa che - utilizzando in entrambi i casi un tempo di evoluzione immaginario $\tau = 8$ e un numero di particelle del polimero $M = 300$ - la precisione del risultato non dipende dallo stato di partenza. Queste considerazioni sono state ottenute esclusivamente dall'osservazione delle distribuzioni di probabilità. Vediamo in seguito quali altre informazioni possiamo acquisire dallo studio dei valori di aspettazione del potenziale sugli stati campionati.

Osservazioni:

Questo risultato è in accordo con le previsioni teoriche dato che la funzione a doppia gaussiana - con i parametri ottenuti dal VMC - è simile alla funzione d'onda di GS. Calcolare il valore di aspettazione del potenziale alle estremità del polimero significa, infatti, calcolare l'elemento di matrice $\langle \psi_0 | \hat V| GS \rangle$ e $\langle GS | \hat V|\psi_0  \rangle$. Quello che ci si aspetta, quindi, è che se lo stato di partenza $|\psi_0\rangle$ è simile al ground state esatto, anche per $\lambda \rightarrow 0$ (o similmente $\lambda \rightarrow 2\tau$) il valore di aspettazione sia simile a quello che effettivamente si ha sul GS;


* la proiezione delle funzioni costante e gaussiana sul GS richiede un'evoluzione per tempi immaginari più lunghi rispetto al caso della doppia gaussiana. Questo accade perchè queste due funzioni differiscono notevolmente - anche se in misura diversa - dalla funzione d'onda di GS. Agli estremi del polimero i valori di aspettazione del potenziale - ovvero gli elementi di matrice $\langle \psi_0 | \hat V| GS \rangle$ e $\langle GS | \hat V|\psi_0  \rangle$ - sono quindi ben diversi dal valore di aspettazione sul GS. In particolare, la gaussiana viene proiettata sul GS per $\lambda > 40$, mentre la costante per $\lambda > 20$. Tuttavia, l'andamento del potenziale è molto diverso: per valori molto piccoli di $\lambda$ la proiezione della costante differisce di più dal GS rispetto alla gaussiana. Una giustificazione teorica all'andamento del potenziale alle estremità del polimero si può ottenere analizzando gli elementi di matrice $\langle \psi_0 | \hat V| GS \rangle$ e $\langle GS | \hat V|\psi_0  \rangle$ se si esprime lo stato $|\psi_0\rangle$ come combinazione lineare degli autostati dell'Hamiltoniana del problema. Se lo stato di partenza ha un overlap non trascurabile con uno o più autostati eccitati dell'Hamiltoniana ($\langle \psi_0 | E_\alpha\rangle$, dove $| E_\alpha\rangle$ è l'autostato $\alpha-esimo$), allora è possibile che questo contributo sia osservabile nell'andamento del valore medio del potenziale agli estremi del polimero.

## PIMC - T > 0

Il PIMC è una tecnica Monte Carlo per lo studio di sistemi quantistici che permette di calcolare i valori di aspettazione di una osservabile $\hat O$ su una miscela statistica di stati distribuiti secondo distribuzione di Boltzmann. La tecnica consiste nel mappare la media $\textit{quantistica}$ $\langle \hat O \rangle=Tr[\hat \rho \hat O]$, dove $\hat\rho=e^{-\beta\hat H}/Tr[e^{-\beta\hat H}]$ è l'operatore statistico canonico e $\beta$ l'inverso della temperatura, in una media canonica $\textit{classica}$ sfruttando la procedura di path integral. 

Se si considera un operatore diagonale $\hat O$ sulla base delle coordinate, la media quantistica a temperatura finita diventa, in rappresentazione delle coordinate (1 particella in 1 dimensione):

$$\langle\hat O\rangle=Tr(\hat\rho)^{-1}\int dx \rho(x,x,\beta)O(x) \tag{5}$$

dove $\rho(x,x,\beta)=\langle x|e^{-\beta\hat H}|x\rangle$ può essere calcolato - grazie alla formulazione path integral - in maniera analoga a quanto fatto nel caso a temperatura nulla con il propagatore $G(x,x',t)=\langle x|e^{-\frac{it}{\hbar}}|x'\rangle$, dove $\beta$ prende il posto del tempo immaginario $\tau=it/\hbar$. Anche in questo caso compare il parametro $M$ che divide $\beta$ e che comporta la comparsa di $M$ integrazioni. Il valore di aspettazione $(5)$ può quindi essere espresso nel modo seguente:

$$\langle\hat O\rangle=\int dx\int dx_2\dotsb\int dx_M\underbrace{\frac{\rho_p(x,x_2,\beta/M)\dotsb\rho_p(x_M,x,\beta/M)}{Tr(\hat\rho)}}_{p(x,x_2,\dotsb,x_M)}O(x)$$ 

dove $\rho_p(x_i,x_{i+1},\beta/M)= (4)$ con $\delta\tau=\beta/M$.
$p(x,x_2,\dotsb,x_M)$, che rappresenta a tutti gli effetti una distribuzione di probabilità ed è quindi campionabile con metodi Monte Carlo, descrive un polimero chiuso di $M$ particelle "classiche" che interagiscono con una forza elastica di origine cinetica in cui la costante elastica $k$ è proporzionale a $M/\beta$. Ciò significa che a basse temperature ($k$ piccolo) le particelle del polimero hanno maggiore possibilità di muoversi (regime quantistico, particella delocalizzata), mentre ad alte temperature ($k$ grande) il polimero collassa sempre di più in un unico punto (regime classico, particella localizzata). 

### Campionamento della densità di probabilità della posizione della particella per $T>0$

Si è scelto di studiare il sistema dei due esercizi precedenti (potenziale $V(x)=x^4 - \frac{5}{2} x^2$ ) per 15 temperature diverse nel range $[0.1,1.5]$. Di seguito sono mostrate le distribuzioni di probabilità campionate con l'algoritmo PIMC. Per la simulazione è stato utilizzato un polimero di 30 elementi - numero sufficiente affinchè la quantità $(\frac{\beta}{M})^2 << 1$ e che quindi l'approssimazione primitiva fosse valida. 

I parametri $BB$ e $\delta$ utilizzati per effettuare le mosse del polimero sono stati scelti in maniera da ottenere un'accettazione pari all'incirca al 50%. I loro valori sono stati quindi cambiati in funzione della temperatura. Si è osservato che il valore dei parametri deve diminuire al decrescere della temperatura: a temperature più basse, infatti, il sistema è meno ergodico.

```python
# plot istogramma distribuzione di probabilità campionata con PIMC per diversi valori della temperatura T

plt.figure(figsize=(12,8))

temp = np.linspace(0.1,1.5,15)
temp =np.append(temp,[2.5,5.0])
alpha = np.linspace(0.05,1,17)

bins,histo,err_histo = np.loadtxt("PIGS/probability_doublegauss.dat",usecols = (0,1,2), unpack = 'true')

# plot distribuzione di ground state campionata con PIGS
plt.errorbar(bins, histo, err_histo, label = "T = 0" , c = 'black')

scale = 0.2
# plot potenziale
plt.plot(x,V*scale, label="Potential") 

for i in range(17):
    bins,histo,err_histo = np.loadtxt("PIMC/probability_T"+str(round(temp[i],1))+".dat",usecols = (0,1,2), unpack = 'true')

    # plot distribuzione campionata con PIMC
    plt.errorbar(bins, histo, err_histo, label = "$T = "+str(round(temp[i],1))+"$",color='green',alpha=alpha[i] )
        
plt.xlabel('Posizione [unità naturali]', size = 12)
plt.xlim(-2.5, 2.5)
plt.ylim((-0.4,0.5))
plt.title('Densità di probabilità per differenti valori di $T > 0$', size = 15)
plt.grid()
plt.legend()
plt.show()
```

All'aumentare della temperatura la distribuzione di probabilità della posizione della particella presenta dei picchi più alti e sempre più centrati nei due minimi del potenziale. Di contro, la densità di probabilità diminuisce nella zona tra i due picchi, dove è presente una barriera di potenziale. I risultati del PIMC sono quindi in accordo con la teoria: con l'aumento della temperatura il sistema quantistico approccia il corrispondente sistema classico. Se la temperatura viene aumentata ulteriormente (ad esempio a $T=5$), quindi, la distribuzione di probabilità tende ad approssimare una distribuzione di probabilità uniforme.
