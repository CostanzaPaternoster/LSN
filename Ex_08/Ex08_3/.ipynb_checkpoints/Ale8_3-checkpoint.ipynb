```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

def gstate(x, a, b):
    return 1/(b**2 * math.pi * (2 + np.exp(-a**2/b**2)))*(np.exp(-0.5*(x-a)**2/b**2) + 
                                                          np.exp(-0.5*(x+a)**2/b**2))**2
def mexhat(x):
    return x**4 - 2.5*x**2


# Soluzione del problema tramite il metodo di discretizzazione presentato nel jupyter-
# notebook dell'esercizio. Questi risultati sono utili per un confronto.

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
xgalli = np.linspace(-a/2, a/2, N)
dx = xgalli[1] - xgalli[0] # the step size
V = mexhat(xgalli)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
```

## 1. Brevi cenni di teoria

Le tecniche PIGS (Path Integral Ground State) e PIMC (Path Integral Monte-Carlo) sono due tecniche Monte-Carlo utilizzate per lo studio di sistemi quantistici a temperatura zero e a temperatura finita, rispettivamente. In entrambe il risultato fondamentale è il $\textit {mapping}$ di un calcolo di un valore di aspettazione "quantistico", ovvero nella tipica forma

$$
\langle O \rangle_{MQ} = Tr[ \hat O \hat \rho ] \tag{1}
$$

dove $\hat O$ è un'osservabile e $\hat \rho $ è l'operatore densità, nel calcolo di un valore atteso $\textit{classico}$ del tipo

$$
\langle O \rangle_{CL} = \int dx \int dy \hspace{0.3 cm}O(\cal x) p(x, y) \tag{2}
$$

dove si è lasciata intenzionalmente sottintesa la dimensionalità del problema in questione.
La comparsa delle variabili di integrazione "classiche" aggiuntive $y$ (e quindi l'aumento della complessità del calcolo), è il prezzo da pagare per poter effettuare questo mapping. A questo punto infatti il problema può essere risolto tramite tecniche Monte-Carlo, campionando la distribuzione di probabilità $p(x, y)$ e calcolando quindi i valori medi  delle osservabili di interesse.

### 1.01 PIGS, ovvero $\hat \rho = |GS\rangle \langle GS|$

Siamo in questo caso interessati al valor medio di una certa osservabile $\hat O$ sul ground state $|GS\rangle $ di una certa hamiltoniana $\hat H$, tipicamente della forma $\hat H = \hat T + \hat V $. Essendo il ground state ignoto, nel PIGS si sfrutta la seguente proprietà dell'evoluzione in tempo immaginario: indicando con $|\psi_0\rangle$ un generico stato del sistema e con $\{ |E_\alpha\rangle \}_{\alpha = 0}^\infty$ la base degli autostati di $\hat H$ con $|E_0\rangle = | GS \rangle$, si ha

$$
|\psi (\tau) \rangle := e^{-\hat H \tau} |\psi_0\rangle =  e^{-\hat H \tau} \sum_{\alpha = 0}^\infty | E_\alpha \rangle \langle E_\alpha |\psi_0 \rangle  = \sum_{\alpha = 0}^\infty c_\alpha e^{- E_\alpha \tau}  |E_\alpha\rangle, \hspace{0.3 cm} \text{con } c_\alpha = \langle E_\alpha|\psi_0\rangle
$$

$$
\frac{|\psi (\tau) \rangle}{\sqrt{\langle \psi (\tau)|\psi (\tau) \rangle}} = \frac{e^{- E_0 \tau}\big( c_0 |GS \rangle + \sum_{\alpha = 1}^\infty c_\alpha e^{- (E_\alpha - E_0)\tau} |E_\alpha\rangle \big)}{e^{- E_0 \tau}\sqrt{c_0^2 + \sum_{\alpha = 1}^\infty c_\alpha^2 e^{- 2(E_\alpha - E_0)\tau}}} \underset{\tau \rightarrow \infty}{\rightarrow} |GS\rangle \tag{3}
$$

ovvero l'evoluzione in tempo immaginario (normalizzata) di uno stato qualsiasi tende per tempi immaginari grandi al ground state. Sfruttando questa proprietà si può riscrivere il valor medio di un'osservabile sul ground state come segue 
$$
\langle GS | \hat O | GS \rangle = \underset{\tau \rightarrow \infty}{\lim}\frac{\langle \psi_0 | e^{-\hat H \tau }\hat O e^{-\hat H \tau }| \psi_0 \rangle}{\langle \psi (\tau)|\psi (\tau) \rangle} . \tag{4}
$$
Facendo ora uso della procedura di path integral e di alcune approssimazioni si può finalmente scrivere questo valor medio in funzione di quantità note. Riportiamo qui direttamente il risultato (ricavato a lezione) di questa procedura per un'osservabile $\hat O$ diagonale sulla base delle coordinate (nell'esercizio si avrà $\hat O = \hat V$, l'energia potenziale):

$$
\langle GS| \hat O |GS \rangle \approx \int \Big( \prod\limits_{i=1}^{2M+1} d\vec r_i \Big) \overbrace{O(\cal \vec r_{\textrm{M+1}})}^{\large{O(x)}} \hspace{0.1 cm} \underbrace{\frac{\psi_0 (\vec r_1) \hspace{0.1 cm}\psi_0 (\vec r_{\textrm{2M+1}})}{\langle \psi (\tau)|\psi (\tau) \rangle} \hspace{0.1 cm} \prod\limits_{i=1}^{\textrm{2M}} G(\vec r_i,\vec r_{i+1}, \frac{\tau }{ \textrm M})}_{\large{p(x, y)}} \tag{5}
$$

Qui la generica varibile $\vec r$ contiene le posizioni delle particelle fisiche considerate ($3N$ variabili) mentre $M$ è un parametro che compare nella procedura del path integral: all'aumentare di $M$ il termine a destra approssima meglio il termine a sinistra, ma il numero di integrazioni di effettuare aumenta come $3N\cdot(2M+1)$. Osserviamo inoltre come compaia esplicitamente $\tau$, poichè una delle approssimazioni consiste proprio nel troncare a un tempo finito l'evoluzione immaginaria nella $(4)$. Infine

$$
G(\vec v, \vec w, \delta \tau) = \big(\frac{1}{4\pi\delta \tau\gamma}\big)^{3/2}exp\big\{{-\frac{\delta \tau}{2} V(\vec v)+\frac{|\vec v-\vec w|^2}{4\delta \tau\gamma}-\frac{\delta \tau}{2} V(\vec w)}\big\} .
$$ 

L'aspetto da mettere in evidenza qui è che la $(5)$, con le identificazioni indicate, è esattamente nella forma $(2)$ e quindi il calcolo può essere svolto come spiegato sopra. Un dettaglio a cui fare attenzione è però il seguente. Nel calcolo dell'integrale $(2)$ possiamo osservare un'asimmetria nelle variabili di integrazione: mentre l'osservabile $\hat O$ dipende solo dalle variabili delle particelle "fisiche" (nel nostro caso le $3N$ posizioni $\vec r_ {M+1}$), la distribuzione di probabilità $p(x, y)$ dipende non solo da queste ultime, ma anche dalle variabili aggiuntive che compaiono nella procedura del path integral (le restanti $3N\cdot2M$ variabili). Queste variabili in più hanno un'interpretazione classica come posizioni di particelle "fittizie" formanti un $\textit{polimero speciale }$  (l'aggettivo "speciale" compare poiché le particelle di questo polimero interagiscono secondo alcune regole particolari). Formalmente questa differenza passa inosservata poiché la procedura genera in maniera automatica, a partire dalla $(4)$, le variabili da associare all'osservabile; se si procede invece prima campionando la distribuzione di probabilità $p(x, y)$ e poi calcolando i valori medi, bisogna "manualemnte" scegliere in questo secondo momento quali variabili inserire in $O$ e questa scelta influenza il risultato. Un'analisi attenta del modo in cui viene ricavata la $(5)$ (vedi lezione) mostra che cambiare le variabili associate a $O$ nell'integrazione, cioè cambiare la posizione lungo il polimero in cui viene valutata l'energia potenziale, corrisponde a calcolare non più il valor medio "simmetrico" $\langle \psi_0 | e^{-\hat H \tau }\hat O e^{-\hat H \tau }| \psi_0 \rangle$ bensì un valor medio "asimmetrico" del tipo $\langle \psi_0 | e^{-\hat H \tau_1 }\hat O e^{-\hat H \tau_2 }| \psi_0 \rangle$ con $\tau_1 + \tau_2 = 2 \tau$. Torneremo dopo su questa osservazione per spiegare alcuni risultati.

## 2. Applicazione: 1D quantum particle in un potenziale esterno 

In questo paragrafo mostriamo un'applicazione del problema precedente ad un caso molto semplice di una singola particella che si muove in 1 dimensione, soggetta ad un potenziale esterno della forma $V(x) = x^4 - \frac{5}{2} x^2$. Siamo interessati al valor medio dell'energia potenziale sul ground state ignoto di questo problema, e vogliamo in particolare studiare come cambiano i risultati al variare della funzione d'onda di prova $\psi_0$. Per l'evoluzione in tempo immaginario si è scelto $\tau = 8.0$. Avendo fissato $\tau$, per il parametro $M$ sono stati scelti valori che garantissero, nella procedura del sampling Metropolis, un'accettazione prossima al 50% e si è trovato che un valore $M = 450$ soddisfaceva questa richiesta in tutti i casi.
Si sono quindi esplorate le 3 seguenti possibilità:

$$
\psi_0(x) =
\left\{
	\begin{array}{ll}
		1   \hspace{9 cm}\text{(prj1)}\\
		exp\big\{-\frac{x^2}{2}\big\} \hspace{7 cm}\text{(prj2)}\\
        exp\big\{-\frac{(x-\mu)^2}{2\sigma^2}\big\} + exp\big\{-\frac{(x+\mu)^2}{2\sigma^2}\big\}\hspace{3 cm}\text{(prj3)}
	\end{array}
\right.
$$

L'ultima funzione è un'approssimazione del ground state esatto, in cui i valori di $\mu$ e $\sigma$ sono stati scelti in base a considerazioni precedenti (vedi esercizio LSN_Exercise_08.1-2).:

$$
\mu = 0.8 \hspace{1 cm}
\sigma = 0.62
$$

### 2.01 Campionamento della distribuzione di probabilità associata al ground state

Come primo risultato mostriamo il campionamento della funzione d'onda del ground state. Si è utilizzato il codice PIGS a aprtire dalle 3 funzioni indicate.

```python
plt.figure(figsize=(12,8))
text = ['const', 'gauss', 'var']

# Plotto le funzioni risultanti dal codice PIGS
for i in range(3):
    pos, hist, err = np.loadtxt("Output_PIGS/probability."+text[i]+".dat", 
                                usecols = (0,1,2), unpack = 'true')
    plt.errorbar(pos, hist, err, label = text[i])
    
# Fit con somma di curve gaussiane come da ipotesi dell'esercizio LSN_Ex08.1-2
p_opt, p_cov = curve_fit(gstate, pos, hist)
y_fit = gstate(pos,p_opt[0],p_opt[1])
plt.plot(pos,y_fit, label='fit somma gaussiane') 

# Funzione d'onda ricavata risolvendo l'eq di Schroedinger con il metodo di 
# discretizzazione
plt.plot(xgalli,(psi[0])**2, label = 'Galli discretized method', c = 'black')

plt.xlabel('Position (natural units)', size = 12)
plt.xlim(-2.5, 2.5)
plt.title('Densità di probabilità', size = 20)
plt.grid()
plt.legend()

print("Parametri ottimali del fit:")
print("\u03BC = ", p_opt[0])
print("\u03C3 = ", p_opt[1])
print("Parametri iniziali della funzione d'onda approssimata:")
print("\u03BC = ", 0.80)
print("\u03C3 = ", 0.62)
```

Osserviamo come il campionamento dia risultati identici in tutti e 3 i casi (curve verdi, blu ed arancione), il che indica la grande potenza del metodo, che converge ad uno stesso risultato, con una stessa confidenza (barre d'errore), indipendentemente dallo stato di partenza. Dal fit (curva rossa) possiamo anche notare come l'assunzione alla base dell'esercizio LSN_Exercise_08.1-2 "variational MC", ovvero che il ground state sia scrivibile come somma di due gaussiane, sia un evidente approssimazione della forma funzionale del ground state vero (dove con "vero" ci si riferisce a quello trovato con il calcolo PIGS). Il risultato che troviamo è invece perfettamente in accordo con quello trovato tramite il metodo di discretizzazione (curva nera) presentato nella consegna dell'esercizio.

### 2.02 Valor medio potenziale sul ground state

A questo punto vogliamo invece trovare un modo di mettere in evidenza le differenze che possono manifestarsi a seconda della diversa funzione d'onda di partenza, differenze che sembrano essere assenti nei grafici appena riportati. Per fare questo, ricordando l'osservazione fatta alla fine del paragrafo $2.$, plottiamo la funzione 

$$
\langle V(\lambda) \rangle = \frac{\langle \psi_0 | e^{-\hat H  \lambda }\hat V e^{-\hat H (2\tau-\lambda) }| \psi_0 \rangle}{\langle \psi(\tau) | \psi(\tau) \rangle}, \hspace{0.5 cm} \lambda \in [0, 2\tau]. \tag{6}
$$ 

dove $\lambda$ reppresenta l'asimmetria nell'evoluzione in tempo immaginario dei due stati. Per quanto detto sopra, cambiare $\lambda$ significa cambiare posizione lungo il polimero speciale: per $\lambda = \tau$ (centro del polimero) avremo il valore medio simmetrico $(5)$ , mentre per $\lambda =0, 2 \tau$ (estremi del polimero) avremo una buona approssimazione degli elementi di matrice $\langle \psi_0 | \hat V| GS \rangle$ e $\langle GS | \hat V|\psi_0  \rangle$. Quello che ci si aspetta è che se lo stato di partenza $|\psi_0\rangle$ è simile al ground state esatto (caso $\text{(prj3)}$), anche per $\lambda \rightarrow 0$ (o similmente $\lambda \rightarrow 2\tau$) il valore di aspettazione sia vicino al valore intermedio $(4)$.

```python
text = ['const', 'gauss', 'var']
mean = np.zeros(3)
f = plt.figure(figsize = (12,5))
for i in range(3):
    
    timeslice, ave_pot, err_pot = np.loadtxt("Output_PIGS/potential."+text[i]+".dat", 
                                usecols = (0,1,2), unpack = 'true')
    for j in range(100, 350):
        mean[i] += ave_pot[j]/250 #timeslice[j].astype(int)
    f.add_subplot(1, 2, 1)
    plt.errorbar(timeslice, ave_pot, err_pot, label = text[i])
    plt.xlabel('\u03BB')
    plt.title('Elemento di matrice del potenziale')
    plt.grid()
    plt.legend()
    
    f.add_subplot(1, 2, 2)
    plt.errorbar(timeslice, ave_pot, err_pot, label = text[i])
    plt.xlim(0,60)
    plt.xlabel('\u03BB')
    plt.title('Elemento di matrice del potenziale (dettaglio)')
    plt.grid()
    plt.legend()


print("Valor medio energia potenziale (media sulle 250 posizioni centrali del polimero):")
print("Proiezione costante, <V> = ", mean[0])
print("Proiezione gaussiana, <V> = ", mean[1])
print("Proiezione somma di gaussiane, <V> = ", mean[2])
```

Il comportamento coincide con quello previsto anche se vale la pena fare alcune osservazioni:
1. sebbene $\text{(prj3)}$ (funzione "variational MC") arrivi più in fretta al valore medio $-0.78$ rispetto a $\text{(prj1)}$ (funzione costante = 1), questa differenza è in questo caso molto piccola, meno di 20 posizioni lungo il polimero i.e. passi di evoluzione in tempo immaginario
2. il caso $\text{(prj2)}$, in cui si parte da una curva gaussiana, decresce più lentamente della costante e giunge al valor medio per $\lambda = 50$.

Questo comportamento è quantomeno curioso: è davvero l'algoritmo PIGS che è in grado di proiettare molto velocemente una qualsiasi funzione sul ground state? E se sì, perchè la funzione costante=1 più velocemnte della gaussiana? Per rispondere a queste domande bisogna capire se il PIGS si comporta così anche per altri sistemi fisici, e quindi si è studiato il caso di una particella, sempre in 1 dimensione, ma soggetta ad un potenziale armonico.
I risultati sono presentati qui sotto.

```python
mean = np.zeros(3)
f = plt.figure(figsize = (12,5))
for i in range(3):
    
    timeslice, ave_pot, err_pot = np.loadtxt("Output_PIGS/harmonic."+text[i]+".dat", 
                                usecols = (0,1,2), unpack = 'true')
    for j in range(100, 350):
        mean[i] += ave_pot[j]/250 #timeslice[j].astype(int)
    f.add_subplot(1, 2, 1)
    plt.errorbar(timeslice, ave_pot, err_pot, label = text[i])
    plt.xlabel('\u03BB')
    plt.title('Elemento di matrice del potenziale')
    plt.grid()
    plt.legend()
    
    f.add_subplot(1, 2, 2)
    plt.errorbar(timeslice, ave_pot, err_pot, label = text[i])
    plt.xlim(0,110)
    plt.xlabel('\u03BB')
    plt.title('Elemento di matrice del potenziale (dettaglio)')
    plt.grid()
    plt.legend()


print("Valor medio energia potenziale (media sulle 250 posizioni centrali del polimero):")
print("Proiezione costante, <V> = ", mean[0])
print("Proiezione gaussiana, <V> = ", mean[1])
print("Proiezione somma di gaussiane, <V> = ", mean[2])
```

La curva arancione rappresenta il caso in cui si parta da una gaussiana come funzione di prova: coincidendo questa con il ground state dell'oscillatore armonico, il potenziale che ne risulta è, indipendentemente da $\lambda$, quello del ground state, ovvero (in unità naturali) $0.25$. La curva verde è invece quella in cui si è proiettata la funzione d'onda del ground state del problema precedente ($\text{(prj3)}$, ovvero una somma di gaussiane). L'aspetto interessante è che in questo caso la funzione costante (curva blu) viene proietta sul ground state per $\lambda > 100$,  un comportamento ben diverso da quello che si aveva nel caso precedente ($\lambda \approx 20$). Cosa è cambiato? La risposta è che agli estremi del polimero diventano importanti effetti dovuti alla sovrapposizione della funzione d'onda di partenza con gli stati eccitati del sistema. In particolare quando calcoliamo l'integrale $(5)$ per $\lambda$ prossimo a $0$ (ma lo stesso vale per $\lambda$ prossimo a $2\tau$), dalla $(6)$ ci accorgiamo che

$$
\langle V(\lambda) \rangle \big\rvert_{\lambda \approx 0} \approx \frac{\langle \psi_0 | e^{-\hat H  \lambda }\hat O | GS \rangle}{\sqrt{\langle \psi(\tau) | \psi(\tau) \rangle}} = \frac{\sum_{\alpha = 0}^\infty \langle \psi_0 | E_\alpha\rangle e^{- E_\alpha  \lambda }\langle E_\alpha|\hat O | GS \rangle}{\sqrt{\langle \psi(\tau) | \psi(\tau) \rangle}} = 
$$
$$
 = \frac{c_0 e^{- \tau E_0}\big(O_{00} + \sum_{\alpha=1}^\infty c_\alpha e^{- \lambda (E_\alpha -E_0) }O_{\alpha 0}\big)}{c_0 e^{- \tau E_0}\sqrt{1 + \sum_{\alpha=1}^\infty c_\alpha^2 e^{-2\tau(E_\alpha - E_0)}}} \approx 
$$
$$
\approx O_{00} + c_1 e^{- \lambda (E_1 -E_0) }O_{1 0} + c_2 e^{- \lambda (E_2 -E_0) }O_{2 0} + ... \tag{7}
$$

dove con $O_{ij}$ si è indicato il valor medio dell'osservabile $\hat O$ sugli autostati $esatti$ del problema. Dunque se per la funzione di partenza ha un overlap $c_\alpha$ non trascurabile con un certo autostato esatto $|E_\alpha\rangle$, e se contemporaneamente l'elemento di matrice del potenziale $\langle E_\alpha|\hat V| GS\rangle$ è diverso da $0$, per $\lambda$ piccolo il valore dell'integrale PIGS $(5)$ conterrà questi termini correttivi. 
Per verificare che effettivamente questi temrini giustificano l'andamento dei grafici è necessaria una conoscenza delle funzioni d'onda esatte del problema. Questo studio è stato fatto (per non allungare troppo questo foglio) nel jupyter_notebook "Effetti correttivi-...", sfruttando lo spettro prodotto dalla procedura di discretizzazione: si è effettivamente verificato che un plot della funzione $(7)$ (in realtà un "raffinamento" di quest'ultima, contenente altri termini correttivi), riproduce esattamente il conto PIGS sia nel caso del potenziale originale che nel caso del potenziale armonico. Inoltre si è verificato che, nel primo caso, gli overlap $c_\alpha$ della funzione costante (dall'ordine $4$ in poi) sono sistematicamente più grandi (fattore 10-100) di quelli della gaussiana, il che spiega la diversa rapidità di "proiezione" delle due funzioni.

## 3. PIMC ovvero $\hat \rho = e^{-\beta \hat H}/\text{Tr}[e^{-\beta \hat H}]$

Quanto detto per il PIGS vale con qualche piccola modifica anche il PIMC. In questo caso si vuole calcolare una media $\textit {termica}$ all'equilibrio di una certa osservabile $\hat O$, ovvero la quantità

$$
\langle \hat O \rangle = \frac{\text{Tr}[e^{-\beta \hat H}\hat O]}{\text{Tr}[e^{-\beta \hat H}]}. \tag{8}
$$

Qui non abbiamo bisogno di utilizzare l'evoluzione in tempo immaginario, e il ruolo giocato dal parametro $\tau$ nel PIGS viene qui interpretato dal parametro $\beta$, l'inverso della temperatura. Sfruttando ancora una volta la procedura del path integral si può scrivere per la $(8)$ una formula analoga alla $(5)$:

$$
\langle \hat O \rangle = \frac{1}{\text{Tr}[e^{-\beta \hat H}]}\int \big(\prod\limits_{i = 1}^M d\vec r_i\big) O(\vec r_1) \underbrace{ \Big(\prod\limits_{i=1}^MG(\vec r_i, \vec r_{i+1}, \frac{\beta}{M}) \Big)\big \rvert_{\;\vec r_{M+1} = \vec r_1}}_{\large p(x, y)} \;. \tag{9}
$$

Osserviamo come nel PIMC, essendo la $p(x, y)$ simmetrica rispetto a tutte le variabili, non si presenti la questione dell'asimmetria nell'integrazione che abbiamo messo in evidenza per il PIGS.

### 3.01 Applicazione al problema della particella

Vogliamo ora campionare la distribuzione di probabilità associata alla particella (istogramma delle configurazioni) al variare della temperatura.

Nel codice PIMC è stato usato un polimero di $30$ particelle e si è esplorato un range di temperature $[0.1, 1.5]$ in $15$ punti equidistanti. 
Riportiamo per chiarezza le regole utilizzate per modificare i parametri del codice affinché l'accettazione del Metropolis fosse prossima al 50%: 

$$
T_n = T_{max} - \frac{n-1}{10}\hspace{1 cm} T_{max} = 1.5
$$

$$
\delta_n = \delta_{max} - \frac{n}{12}\hspace{1 cm} \delta_{max} = 1.8
$$

$$
bb_n = 20 - 5\times(n - 12.5)
$$

Qui $n$ indica l'n-esima simulazione ($n$ = 1,..., 15), $T_n$ la temperatura, $\delta_n$ il parametro delle traslazioni e $bb_n$ il numero di ricostruzioni del polimero ($\textit{BrownianBridgeReconstruction}$). 

Queste regole sono "empiriche", costruite $\textit{ad hoc}$ per garantire l'accettazione desiderata, e i parametri numerici che compaiono non hanno quindi alcun significato preciso e possono sicuramente essere cambiati. Quello che invece $\textit{ha}$ significato è come i parametri cambiano rispetto alla temperatura: come regola generale si è cioè trovato che, al diminuire di $T$, anche $\delta_n$ e $bb_n$ devono diminuire per garantire l'accettazione ~50%. Questo è in accordo con il fatto che, a temperature basse, lo spazio "esplorabile" è ridotto. 

Si nota infine che $bb_n$ è, per diversi $n$, $> 30$, cioè maggiore della lunghezza del polimero. Questo fatto è possibile grazie alla presenza delle PBC: una ricostruzione più lunga del polimero stesso è una ricostruzione che gira più volte lungo il polimero, aumentando quindi l'ergodicità.

Vediamo ora i risultati.

```python
files = 15
plt.figure(figsize=(10,8))

for i in range(1, files+1):
    with open("Output_PIMC/probability."+str(i)+".dat") as f:
        temp = f.readline(20)
    rang, prob, err = np.loadtxt("Output_PIMC/probability."+str(i)+".dat", 
                                 usecols = (0,1,2),skiprows = 3,  unpack = 'true')
    plt.errorbar(rang, prob, err, label = str(temp), c = str(i**2+20))

pos, hist, err = np.loadtxt("Output_PIGS/probability.var.dat", 
                               usecols = (0,1,2), unpack = 'true')
plt.errorbar(pos,hist, err, label = 'Temperature 0', c = 'r')

plt.xlabel('x', size = 12)
plt.title('Istogramma configurazioni per diverse T', size = 15)

plt.xlim(-2.5, 2.5)
plt.grid()
plt.legend()
```

Le configurazioni della particella possono quindi essere descritte in questo modo.

1. $T \rightarrow 0$ (colore scuro, la curva rossa è quella del PIGS a $T = 0$): la distribuzione di probabilità si restringe lateralmente e si abbassano i due massimi, mentre il minimo centrale cresce $\Rightarrow$ la particella è piu delocalizzata, e la probabilità di trovarla in un minimo (del potenziale) non è molto diversa dalla probabilità di trovarla nella zona centrale "classicamente proibita" (effetto tunnel);

2. $T \rightarrow \infty$ (colore chiaro): la distribuzione di probabilità si alza nei massimi e si schiaccia nel minimo $\Rightarrow$ la particella è localizzata, si trova con alta probabilità in uno dei due minimi e difficilmente nella zona "classicamente proibita".
