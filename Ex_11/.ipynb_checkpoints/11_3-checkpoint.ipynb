### Ex. 11.3

In questo esercizio Ã¨ stata costruita ed allenata una rete neurale con l'obbiettivo di fittare una funzione trigonometrica a due variabili: $f(x,y) = \sin(x^2+y^2)$ nel range $x \in [-3/2,3/2]$ e $y \in [-3/2,3/2]$.

Anche in questo caso valgono le considerazioni fatte nell'esercizio precedente. La rete, in questo caso, riceve in ingresso due variabili (x e y) e restituisce una variabile di output z.
Parametri utilizzati:
- $N_{train} = 5000$
- $N_{epochs} = 200 $
- $N_{neurons} = 80$
- $N_{hiddenlayers} = 1$
- ofunzione di attivazione: stochastic gradient descent
- funzione di costo: mse
- funzione di attivazione: relu

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# libraries to compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import SGD, Adagrad, Adadelta
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
```

```python
# functions for generating data: linear, cubic polynomial and 2D trigonometric
def linear(x,m,q):
    return x*m + q
def poly3(x,a,b,c,d):
    return a + b*x + c*x*x + d*x*x*x
def trig2D(x,y):
    return np.sin(x*x + y*y)

# data generation: labeled data for training and validation
def generate_data(function,sigma,lim,dim):
        dim_train = int(dim)
        dim_valid = int(dim/9.)
        x_train = np.random.uniform(-lim, lim,dim_train)
        x_valid = np.random.uniform(-lim, lim,dim_valid)
    
        if function == "linear":
            x_valid.sort()
            y_target = linear(x_valid,2,1)
            y_train = np.random.normal(linear(x_train,2,1), sigma) 
            y_valid = np.random.normal(linear(x_valid,2,1), sigma)
            return x_train,y_train,x_valid,y_valid,y_target
        
        elif function == "poly3":
            x_valid.sort()
            y_target = poly3(x_valid,4,-3,-2,3)
            y_train = np.random.normal(poly3(x_train,4,-3,-2,3), sigma) 
            y_valid = np.random.normal(poly3(x_valid,4,-3,-2,3), sigma)
            return x_train,y_train,x_valid,y_valid,y_target
        
        elif function == "trig2D":
            x_train = np.random.uniform(-lim,lim,dim_train) # x-coordinate
            y_train = np.random.uniform(-lim,lim,dim_train) # y-coordinate
            x_valid = np.random.uniform(-lim,lim,dim_valid)  
            y_valid = np.random.uniform(-lim,lim,dim_valid)
            
            z_train = trig2D(x_train,y_train)
            z_valid = trig2D(x_valid,y_valid)
            
            # Zip the two vectors (using numpy.dstack) to have a single vector of (x, y) points
            # This is necessary since the NN training function requires input in this form
            train = np.dstack((x_train, y_train)).reshape(dim_train,2)
            valid = np.dstack((x_valid, y_valid)).reshape(dim_valid,2) 
            
            return train, valid, x_train,y_train,z_train,x_valid,y_valid,z_valid

# NN creation        
def create_NN(name):
    # instantiate model
    NN = tf.keras.Sequential()  
    if name == 'linear':
        NN.add(Dense(1, input_shape=(1,)))
        
    elif name == 'poly3':
        # Add a hidden layer with 80 neurons
        NN.add(Dense(80, input_shape=(1,), activation = 'tanh'))
        NN.add(Dense(1))

    elif name == 'trig2D':
        NN.add(Dense(10, input_shape=(2,), activation = 'tanh'))
        NN.add(Dense(1))
   
    return NN  
```

```python
# trigonometric function z = sin(x^2 + y^2)
# labeled data: [(x,y),z]

train,valid,x_train,y_train,z_train,x_valid,y_valid,z_valid = generate_data("trig2D",0,1.5,5000)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(x_valid, y_valid, z_valid,
                color = 'b', alpha = 0.4)
ax.scatter(x_valid, y_valid, z_valid,
           color='darkorange', alpha = 0.8,label='labeled data')


plt.grid(True)
plt.legend()
plt.show()
```

```python
# creazione del modello sequenziale
model3 = create_NN('trig2D')

# compilazione: scelta optimizer, loss function e metrics
model3.compile(optimizer='sgd', loss='mse', metrics=['mse'])

model3.summary()
```

```python
# allenamento della rete
history3 = model3.fit(x=train, y=z_train, 
          batch_size=32, epochs=100, verbose =0,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(valid, z_valid))
```
