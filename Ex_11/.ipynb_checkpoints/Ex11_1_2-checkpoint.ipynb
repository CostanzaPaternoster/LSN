### Overview 
In questi esercizi sono state allenate reti neurali del tipo feedforward, utilizzando la libreria Keras, per effettuare delle regressioni su dati (eventualmente affetti da rumore gaussiano). In particolare, sono state realizzate reti neurali per approssimare tre tipi du funzioni:
- funzione lineare: $f(x)=2x+1$
- polinomio di terzo grado: $f(x)=4-3x-2x^2+3x^3$ per $x \in [-1,1]$.
- funzione trigonometrica a 2 variabili: $f(x,y) = \sin(x^2+y^2)$ nel range $x \in [-3/2,3/2]$ e $y \in [-3/2,3/2]$.

L'obiettivo principale è stato prendere confidenza con le reti neurali e la loro struttura, e capire da quali parametri (e in che misura) dipendono le loro performance.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# libraries to compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import SGD, Adagrad, Adadelta
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
```

```python
# functions for generating data: linear, cubic polynomial and 2D trigonometric
def linear(x,m,q):
    return x*m + q
def poly3(x,a,b,c,d):
    return a + b*x + c*x*x + d*x*x*x
def trig2D(x,y):
    return np.sin(x*x + y*y)

# data generation: labeled data for training and validation
def generate_data(function,sigma,lim,dim):
        dim_train = int(dim)
        dim_valid = int(dim/9.)
        x_train = np.random.uniform(-lim, lim,dim_train)
        x_valid = np.random.uniform(-lim, lim,dim_valid)
    
        if function == "linear":
            x_valid.sort()
            y_target = linear(x_valid,2,1)
            y_train = np.random.normal(linear(x_train,2,1), sigma) 
            y_valid = np.random.normal(linear(x_valid,2,1), sigma)
            return x_train,y_train,x_valid,y_valid,y_target
        
        elif function == "poly3":
            x_valid.sort()
            y_target = poly3(x_valid,4,-3,-2,3)
            y_train = np.random.normal(poly3(x_train,4,-3,-2,3), sigma) 
            y_valid = np.random.normal(poly3(x_valid,4,-3,-2,3), sigma)
            return x_train,y_train,x_valid,y_valid,y_target
        
        elif function == "trig2D":
            x_train = np.random.uniform(-lim,lim,dim_train) # x-coordinate
            y_train = np.random.uniform(-lim,lim,dim_train) # y-coordinate
            x_valid = np.random.uniform(-lim,lim,dim_valid)  
            y_valid = np.random.uniform(-lim,lim,dim_valid)
            
            z_train = trig2D(x_train,y_train)
            z_valid = trig2D(x_valid,y_valid)
            
            # Zip the two vectors (using numpy.dstack) to have a single vector of (x, y) points
            # This is necessary since the NN training function requires input in this form
            train = np.dstack((x_train, y_train)).reshape(dim_train,2)
            valid = np.dstack((x_valid, y_valid)).reshape(dim_valid,2) 
            
            return train, valid, x_train,y_train,z_train,x_valid,y_valid,z_valid

# NN creation        
def create_NN(name):
    # instantiate model
    NN = tf.keras.Sequential()  
    if name == 'linear':
        NN.add(Dense(1, input_shape=(1,)))
        
    elif name == 'poly3':
        # Add a hidden layer with 80 neurons
        NN.add(Dense(80, input_shape=(1,), activation = 'tanh'))
        NN.add(Dense(1))

    elif name == 'trig2D':
        NN.add(Dense(80, input_shape=(2,), activation = 'relu'))
        NN.add(Dense(1))
   
    return NN  
```

### Exercise 11.1

I dati $(x_{i},y_{i})$ utilizzati per il fit sono stati creati a partire dalla funzione $f(x)=2x+1$, a cui è stato aggiunto un rumore $\eta_{i}$ di tipo gaussiano ($N(0,\sigma)$): $y_{i} = 2x_{i} + 1 + \eta_{i}$.

Nel caso della regressione lineare è sufficiente utilizzare un neurone, ovvero 2 parametri (peso $w$ e bias $b$). L'output di questa rete neurale è, infatti, $wx + b$, dove x è il dato di input. In questo semplice caso la performance del modello può essere migliorata aumentando il set di dati utilizzati per l'allenamento della rete, mentre un aumento del numero di neuroni o dei layer complicherebbe inutilmente il modello e peggiorerebbe la sua capacità predittiva. Un altro parametro che può incidere sull'errore di input/output è il numero di epoche $N_{epochs}$, ovvero il numero di ripetizioni di allenamento della rete su uno stesso set di dati. Aumentando $N_{epochs}$ l'errore sul set di training diminuisce: la rete diventa un modello sempre più raffinato per i dati di training. Incrementare il numero di epoche migliora anche l'errore sui dati di test se si ha l'accortezza di non esagerare. Il rischio è - infatti - quello di overfittare i dati di training a discapito della capacità predittiva del modello. 
 
Aumentando il numero di dati di training $N_{train}$ (da 100 a 2000) - mantenendo costante $N_{epochs} = 40$ e il rumore $\sigma = 0.1$ - si è effettivamente osservata una diminuzione degli errori di input/output (da $0.02$ a $0.009/0.01$) e valori dei parametri della rete sempre più vicini a quelli della funzione da cui sono stati generati i dati. Ciononostante, per più di 1000 dati di training gli errori non diminuiscono in maniera significativa, mentre, ovviamente, il tempo di allenamento aumenta. 
Per ottenere risultati migliori, la batch size è stata adattata al numero di dati di training. Ad esempio -infatti - una batch size pari a 32 per un dataset contenente 100 dati di training porta ad errori significativamente più grandi ($1/0.1$).

Aumentare $N_{epochs}$ oltre questo valore non comporta un miglioramento degli errori - che si ottiene solo aumentando il numero di dati di training - ma solo un aumento del tempo di allenamento.

La $\sigma$ del rumore, utilizzata per generare i dati, comporta chiaramente maggiori errori di training/test: in assenza di rumore la loss function raggiunge valori dell'ordine ~ $ 10^{-6}$, mentre già per $\sigma = 0.1$ la loss function arriva, nella migliore delle ipotesi, a ~ $10^{-3}$. $1000$ dati di training nel range [-1,1] permettono comunque di ottenere una buona stima dei parametri $w$ e $b$ anche per valori di $\sigma$ ~ $ 0.8$ e una buona predizione di nuovi dati - sempre in quel range.

Per il caso di dati ottenuti con una funzione lineare con aggiunta di rumore gaussiano ($\sigma = 0.1$) si è scelto quindi di utilizzare un numero di dati di training $N_{train} = 1000$ e un numero di epoche $N_{epochs} = 40$. Il numero di dati di validazione $N_{valid}$ è stato scelto sempre pari al $10$ % del numero di dati totali.

```python
# linear function y = 2x + 1
# labeled data: (x,y + random noise)

np.random.seed(0)

x_train,y_train,x_valid,y_valid,y_target = generate_data("linear",0.1,1,1000)

plt.plot(x_valid,y_valid,'o',color='r',label='labeled noisy data')
plt.plot(x_valid,y_target,label='ideal linear function')
plt.grid(True)
plt.legend()
plt.show()
```

#### Creation of a Sequential model

```python
# create the NN architecture
model = create_NN('linear')
```

#### Compilation
* Metodo di ottimizzazione: Stochastic Gradient Descent (sgd) method
* funzione di costo: Mean Square Error (mse)
* misura della performance (metric): mse 

```python
# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

#### Summary

```python
# get a summary of our composed model
model.summary()
```

#### Training
La dimensione di ciascun batch - il sottoinsieme dei dati necessario per il calcolo della funzione di costo e l'aggiornamento dei parametri - è stata mantenuta costante in tutti i casi ($L_{batch} = 32$).

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=40,
          verbose = 0,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

#### Results

```python
# return weights
w,b = model.get_weights()
print('Parametri ottimizzati della rete: ','\nw = ',w[0],'\nb = ',b)
```

#### Validation and Error Evaluation

```python
# evaluate model
score = model.evaluate(x_train, y_train, batch_size=10, verbose=0)

# print performance
print('--- Errore sui dati di training ---')
print('')
print('Train loss:', score[0])
print('Train accuracy:', score[1])
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=0)

# print performance
print('--- Errore sui dati di test ---')
print('')
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

#### Training History

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid()
plt.show()
```

#### Prediction

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r',label='NN predicted points')
plt.plot(x_valid, y_target,label='ideal linear function')
plt.legend()
plt.grid(True)
plt.show()
```

### Exercise 11.2

In questo esercizio si è cercato di ottenere una rete per l'approssimazione del polinomio di terzo grado $f(x)=4-3x-2x^2+3x^3$ per $x \in [-1,1]$. Per fare questo, si sono esplorate diverse combinazioni di:


- numero di layers
- numero di neuroni in ciascun layer
- funzione di attivazione
- metodo di ottimizzazione
- funzione di costo


In questo caso, dove la funzione da approssimare è un polinomio di terzo grado (4 parametri), la struttura della rete non è più semplice da individuare come nel caso lineare e, inoltre, non è neanche univoca. Il teorema di approssimazione universale, però, ci garantisce che una rete neurale avente un solo layer nascosto può approssimare - con la precisione che si desidera - una qualunque funzione continua. Poichè ciascuna coppia di neuroni permette di costruire una funzione semplice, incrementare il numero di neuroni $N_{neurons}$ in un layer significa migliorare la precisione del modello. E' però importante tenere a mente che aumentare in maniera indiscriminata il numero di parametri della rete, e quindi la complessità del modello, può portare ad un eccessivo adattamento della rete ai dati di training e ad un peggioramento delle performance del modello su un nuovo set di dati. In definitiva si potrebbe ottenere l'indesiderato effetto di incrementare l'errore di output (caso di "low-bias" e "high-variance").

Si è deciso innanzitutto di testare reti con 1 solo hidden layer e un numero variabile di neuroni. Con un set di 5000 dati di training, $N_{epochs} = 200$,funzione di attivazione 'tanh', metodo di ottimizzazione 'sgd' (e batch size 50) e loss function 'mse', le loss function di training e di test (nell'intervallo [-1,1]) raggiungono un ordine di grandezza di $10^{-3}$ già per $N_{neurons}=5$ e non diminuiscono aumentando ulteriormente il numero di neuroni. Il modello sembrerebbe quindi complesso a suffcienza già per 5 neuroni se si considerano dati appartenenti esclusivamente all'intervallo in cui la rete è stata allenata. Al di fuori del range [-1,1], invece, la capacità predittiva della rete con 5 neuroni si è rivelata piuttosto scarsa ($E_{out_{[1,1.5]}} = 3$). Incrementando $N_{neurons}$ la loss function per dati nel range [1,1.5] diminuisce fino a raggiungere il valore $E_{out_{[1,1.5]}} = 0.4$ per $N_{neurons}=200$. Aumentando ulteriormente il numero di neuroni la rete peggiora la sua capacità predittiva sia all'interno del range [-1,1], sia al di fuori di esso. Per $N_{neurons} = 1000$ l'errore sui dati di test nell'intervallo di training raggiunge un ordine di $10^{-2}$ e l'errore per dati nell'intervallo [1,1.5] aumenta a $0.7$. 


Provando a distrbuire i neuroni su 2 hidden layer si è osservato che l'errore sui dati appartenenti al range di training rimane più o meno inalterato rispetto al caso con 1 solo layer,ma di contro, aumenta di molto il tempo di allenamento. Inoltre, aggiungere 1 layer peggiora la capacità del modello di predirre l'andamento del polinomio al di fuori del range di training. 
È possibile che, per il modo in cui funziona l'algoritmo di backpropagation, se si aumenta il numero di layer sia necessario allungare il tempo di training (ovvero il numero di epoche) per ottenere una buona accuratezza del modello. Più il modello si complica, e l'informazione sull'errore deve propagarsi attraverso più layers, più aumenta il rischio di non riuscire ad allenare in maniera adeguata la rete, che necessita di più dati di training. 
In conclusione, 1 unico layer nascosto sembra essere la scelta migliore per questo tipo di problema.

Altri parametri da cui dipende il comportamento di una rete neurale e che è possibile modificare sono:
* la funzione di attivazione: ciascun neurone produce un output a partire dai dati che riceve in ingresso (ovvero una combinazione lineare degli output dei neuroni appartenenti al layer precedente) applicando una specifica funzione - detta di attivazione - a questi ultimi. Esistono varie tipologie di funzioni di attivazione, ma in genere sono caratterizzate da un andamento del tipo on/off: il neurone si attiva solo se l'output supera un certo valore soglia. Esempi di funzioni di attivazione sono la theta di Heaviside (che però può causare problemi a causa della sua discontinuità) o funzioni continue come il sigmoide o la tangente iperbolica. In questo caso è stata scelta la funzione di attivazione 'tanh' perchè è quella con cui la rete si adatta meglio sia con i dati nel range di training che con i dati al di fuori del range di training.
* il metodo di ottimizzazione: è il cuore del machine learning e consiste nel metodo con cui si effettua la ricerca nello spazio dei possibili modelli (che in questo caso sono le reti neurali). In pratica, è il metodo con cui si ottimizza la funzione di costo. E' stato scelto il metodo (mini batch) Stochastic Gradient Descent, ma anche metodi più raffinati, come Adagrad o Adadelta, hanno dato buoni risultati.
* la funzione di costo: è una funzione che permette di quantificare la performance dei modelli candidati. Il modello migliore è quello che ottimizza questa funzione. Nel caso in cui l'output della rete, come in questo caso, è un numero reale, in genere si utilizza lo scarto quadratico medio (mse), di cui si è fatto uso anche in questo caso.

La rete mostrata in seguito è costituita da 1 hidden layer di 80 neuroni aventi funzione di attivazione 'tanh' ed è stata allenata per 200 epoche su un set di 5000 dati con metodo di ottimizzazione 'sgd' (batch size 50) e funzione di costo 'mse'. 

```python
# cubic polynomial y = 4 + -3x - 2x^2 + 3x^3
# labeled data: (x,y+noise)
x_train,y_train,x_valid,y_valid,y_target = generate_data("poly3",0,1,5000)

plt.plot(x_valid,y_valid,'o',color='r',label='labeled data')
plt.plot(x_valid,y_target,label='ideal cubic polynomial')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# create the NN architecture
model = create_NN('poly3')
# compile the model: choose optimizer, loss function e metrics
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

model.summary()
```

```python
# train the net
history = model.fit(x=x_train, y=y_train, 
          batch_size=50, epochs=200, verbose = 0,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# evaluate the model on training data
score = model.evaluate(x_train, y_train, batch_size=50, verbose=0)

# print performance
print('--- Errore sui dati di training ---')
print('')
print('Training loss:', score[0])
print('Training accuracy:', score[1])
```

```python
# valutazione del modello sui dati di validation
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=0)

# print performance
print('--- Errore sui dati di test (intervallo [-1,1]) ---')
print('')
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# training history
# Plot delle funzioni di costo di training e di test
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid()
plt.show()
```

```python
# analisi della predittività della rete nel range [-1,1]
f = plt.figure(figsize = (12,5))

x_pred = np.random.uniform(-1, 1, 100)
y_pred = model.predict(x_pred)
f.add_subplot(1, 2, 1)
plt.scatter(x_pred, y_pred,color='r',label='NN predicted points in the [-1,1] range')
plt.plot(x_valid, y_target,label='ideal polynomial function')
plt.legend()
plt.grid(True)
```

```python
# analisi predittività della rete al di fuori del range [-1,1]
x_pred = np.random.uniform(1, 1.5, 100)
x_pred.sort()
y_pred = model.predict(x_pred)
y_true = poly3(x_pred,4,-3,-2,3)

plt.scatter(x_pred, y_pred,color='r',label='NN predicted points in the [1,1.5] range')
plt.plot(x_pred, y_true,label='ideal polynomial function')

plt.legend()
plt.grid(True)
plt.show()
```

```python
# valutazione del modello sui dati di validation
score = model.evaluate(x_pred, y_true, batch_size=32, verbose=0)

# print performance
print('--- Errore sui dati di test (intervallo [1,1.5]) ---')
print('')
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
