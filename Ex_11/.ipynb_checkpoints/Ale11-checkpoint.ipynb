## Noodling with Neural Networks

In questo esercizio commenteremo quanto scoperto sulle reti neurali del tipo $\textit{Feed Forward}$ utilizzando i pacchetti della libreria Tensorflow tramite l'API Keras.


```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

# Defining fitting functions
def cubic_pol(x, a, b, c, d):
    return a*x*x*x + b*x*x + c*x + d 
def linear(x, m, q):
    return m*x + q
def Sup2D(x, y):
    return np.sin(x*x + y*y)

# Defining some useful functions for the NN
def generate_data(name, interval):
    if name == 'linear':
        Ntraining = 1000    # Number of training points
        Nvalidation = 100   # Number of validation points
        train = np.random.uniform(-interval, interval, Ntraining) 
        valid = np.random.uniform(-interval, interval, Nvalidation)
        valid.sort()
        #sigma = 0.0 # noise standard deviation, for the moment it is absent
        #train = np.random.normal(linear(train, 2, 1), sigma) # actual measures from which we want to guess regression parameters
        #valid = np.random.normal(linear(valid, 2, 1), sigma)
        return train, valid, linear(train, 2, 1), linear(valid, 2, 1)
    
    if name == 'polynomial':
        Ntraining = 5000
        Nvalidation = 500
        train = np.random.uniform(-interval, interval, Ntraining) 
        valid = np.random.uniform(-interval, interval, Nvalidation)  
        valid.sort()
        return train, valid, cubic_pol(train, 3, -2, -3, 4), cubic_pol(valid, 3, -2, -3, 4)
    
    if name == 'surface':
        Ntraining = 10000
        Nvalidation = 500
        train1 = np.random.uniform(-interval, interval, Ntraining) # x-coordinate
        train2 = np.random.uniform(-interval, interval, Ntraining) # y-coordinate
        valid1 = np.random.uniform(-interval, interval, Nvalidation)  
        valid2 = np.random.uniform(-interval, interval, Nvalidation)  
        # Zip the two vectors (using numpy.dstack) to have a single vector of (x, y) points
        # This is necessary since the NN training function require input in this form
        train = np.dstack((train1, train2)).reshape(Ntraining,2)
        valid = np.dstack((valid1, valid2)).reshape(Nvalidation,2) 
        return train, valid, train1,train2,valid1,valid2, Sup2D(train1, train2), Sup2D(valid1, valid2)
    
    
def create_NN(name):
    # instantiate model
    NN = tf.keras.Sequential()
    if name == 'linear':
        # NN composed of just one neuron
        NN.add(Dense(1, input_shape=(1,)))
      
    elif name == 'polynomial':
        # Add a hidden layer with 80 neurons
        NN.add(Dense(80, input_shape=(1,), activation = 'tanh'))
        NN.add(Dense(1))

    elif name == 'surface':
        NN.add(Dense(80, input_shape=(2,), activation = 'relu'))
        NN.add(Dense(1))
   
    return NN
```


### Fit lineare

Un rete con un singolo neurone e senza alcun tipo di funzione di attivazione può essere utilizzata in maniera naturale per fittare una retta. Questo poichè l'output di un singolo neurone è

$$
x \rightarrow w x + b
$$

e quindi facendo i minimi quadrati la "rete" riesce a trovare $w$ e $b$ che identificano la retta che vogliamo fittare. In particolare nell'esercizio il modo per migliorare il risultato della rete con un neurone è quello di aumentare il numero di punti su cui viene allenata. Cambiando la cardinalità del training set da $100$ a $1000$, a parità di epoche di training ($= 30$) i risultati del fit sono ottimi, con una loss function ~$10^{-6}$. 

Abbiamo anche provato a tener fisso il numero dei training points e ad aumentare il numero di epoche: per 200 epoche di training e $100$ training points la loss function, seppur molto bassa ~$10^{-5}$ non ha ancora raggiunto i livelli del caso precedente.

Mostriamo quindi i risultati di questa rete, con $N_{train} = 10^3$, $N_{valid} = 10^2$ e $N_{epochs} = 30$. 

```python
np.random.seed(0)

# Generate training inputs (function_to_fit, extrema_of_interval)
# The TYPE variable is used to reduce the "commenting" throughout the code 
# Line
TYPE = 'linear'
x_train, x_valid, y_train, y_valid = generate_data(TYPE, 1)

# Create the network
model = create_NN(TYPE)

# Compile the network
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# Get a summary of the NN
model.summary()
```

```python
# Train the network with training parmeters
btsize = 32
time = 30
   
history = model.fit(x = x_train, y=y_train, 
          batch_size=btsize, epochs=time,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid),
                   verbose = False)

print('Training completed!')
```

```python
# evaluate model
score = model.evaluate(x_train, y_train, batch_size=50, verbose=0)

# print performance
print("Accuratezza sui dati di training")
print('Test loss:', score[0])

# evaluate model with the exact curve
score = model.evaluate(x_valid, y_valid, batch_size=50, verbose=0)

# print performance
print("\n Accuratezza sui dati di validation")
print('Test loss:', score[0])

# look into training history
# Plot training & validation loss values
plt.figure()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlim(0, 20)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid()
plt.show()
```

```python
# Show actual predictions of the network
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)

# Adjust model output shape to match plot-requirings
dim = len(y_predicted)
y_predicted = y_predicted.reshape(dim,)

# PLot 2D

fig = plt.figure()

plt.scatter(x_predicted, y_predicted,color='r', label = 'NN predicted points')
plt.plot(x_valid, y_valid, label = 'Exact curve')
plt.legend()
plt.grid(True)
plt.show()
```

### Fit del polinomio $y = 3x^3 - 2x^2 - 3x +4$ nel range $x \in [-1, 1]$

In questo caso ci siamo basati sull'argomento presentato a lezione che una rete neurale con un hidden layer è in grado di fittare qualsiasi funzione, grazie all' $\textit{approssimazione tramite funzioni semplici}$. Il punto qui è quindi trovare quanto deve essere densa questa rete, ovvero quanti neuroni deve avere l'hidden layer per approssimare bene il polinomio nel range richiesto.

Abbiamo esplorato varie combinazioni prima di approdare a quella attuale, che consiste di una rete con un hidden layer di 80 neuroni con attivazione "tanh", da allenare su 5000 punti per un numero di epoche superiore a 200. Una configurazione di questo tipo permette però di giungere accuratezze maggiori.

Inizialmente abbiamo provato una rete con 27 neuroni, e ci siamo accorti che potevamo diminuirne il numero fino ad arrivare a 5, senza cambiare di molto l'accuratezza del fit. Questo fit era abbastanza preciso in quasi tutto il range, ma nei pressi di $x = 1$, dove la funzione ha un rapido cambio di direzione, falliva completamente. Abbiamo provato anche ad aggiungere layer ma non il risultato non cambiava. La soluzione è stata infine appunto quella di aumentare il numero di neuroni a 80. Diminuendo questi si vede chiaramente come il fit peggiori nella zona vicina a 1. Viceversa aumentare il numero di neuroni oltre $80$ non cambiava apprezzabilmente il risultato. 

Anche la scelta della funzione di attivazione "tanh" è avvenuta dopo vari tentativi con diverse funzioni, ed è quella che si è rivelata migliore.

```python
# Generate training and validation datasets
# Polynomial
TYPE = 'polynomial'
x_train, x_valid, y_train, y_valid = generate_data(TYPE, 1)

# Create the network
model = create_NN(TYPE)

# Compile the network
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# Get a summary of the NN
model.summary()
```

```python
# Train the network with training parmeters
btsize = 50
time = 200

history = model.fit(x = x_train, y=y_train, 
          batch_size=btsize, epochs=time,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid), 
                   verbose = False)

print('Training completed!')
```

```python
# evaluate model
score = model.evaluate(x_train, y_train, batch_size=50, verbose=0)

# print performance
print("Accuratezza sui dati di training")
print('Test loss:', score[0])

# evaluate model with the exact curve
score = model.evaluate(x_valid, y_valid, batch_size=50, verbose=0)

# print performance
print("\n Accuratezza sui dati di validation")
print('Test loss:', score[0])

# look into training history
# Plot training & validation loss values
plt.figure()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlim(0, 100)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid()
plt.show()
```

```python
# Show actual predictions of the network
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)

# Adjust model output shape to match plot-requirings
dim = len(y_predicted)
y_predicted = y_predicted.reshape(dim,)

# PLot 2D

fig = plt.figure()

plt.scatter(x_predicted, y_predicted,color='r', label = 'NN predicted points')
plt.plot(x_valid, y_valid, label = 'Exact curve')
plt.legend()
plt.grid(True)
plt.show()
```
