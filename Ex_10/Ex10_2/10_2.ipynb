# Ex. 10.2 - Simulated Annealing in parallelo

In questo esercizio è stato risolto il problema del commesso viaggiatore parallelizzando il codice del simulated annealing su 4 core grazie alle librerie di MPI. Poichè il tempo di simulazione è finito, è possibile che con il simulated annealing si giunga solo in prossimità della soluzione ottimale (minimo globale). Parallelizzare permette quindi di eseguire contemporaneamente più simulazioni - utilizzando semi diversi per l'inizializzazione del generatore di numeri casuali - di confrontare i risultati trovati e scegliere tra essi la soluzione ottimale.

Per ciascun core sono state effettuati $1000$ passi di simulated annealing, cioè $1000$ simulazioni per temperature da da $\beta=0$ a $\beta=500$. Le mosse sono state effettuate scegliendo tra scambio tra due città (pair permutation), scambio tra due gruppi di città (group permutation), shift di un gruppo di città, inversione dell'ordine delle città in un gruppo. Ciascuna mossa è stata scelta con probabilità pari al $20\%$, mentre con probabilità $20\%$ non è stata effettuata alcuna mossa.

Di seguito sono mostrati i plot dei cammini più brevi trovati confrontando i risultati finali ottenuti con il simulated annealing in ciascun core, per due disposizioni diverse di 30 città:

- disposizione casuale sulla circonferenza di raggio unitario
- disposizione casuale nel quadrato di lato $[0,1)x[0,1)$

## Plot del percorso più breve

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
 
x1,y1 = np.loadtxt('circle.start',usecols = (0,1), delimiter = ' ', unpack = 'true')
x2,y2 = np.loadtxt('square.start',usecols = (0,1), delimiter = ' ', unpack = 'true')
X = [x1,x2]
Y = [y1,y2]
x = np.zeros(31)
y = np.zeros(31)

file = ['Circle','Square']
mappa = ['Cerchio','Quadrato']
fitness = [0,0]

g = plt.figure(figsize = (13,6))

for i in range(2):
    ID = np.loadtxt(file[i]+'/BestPath.dat',usecols = (0), unpack = 'true')
    ID = np.array(ID, dtype=int)
    fitness[i] = np.loadtxt(file[i]+'/BestFitness.dat',usecols = (0), unpack = 'true')

    for j in range(30):
        x[j] = X[i][ID[j]]
        y[j] = Y[i][ID[j]]

    x[30] = X[i][ID[0]]
    y[30] = Y[i][ID[0]]

    g.add_subplot(1,2,i+1)
    plt.plot(x,y,'-',color='lightblue',linewidth=2.5)
    plt.plot(x,y,'o',color='blue')
    plt.xlabel('x',size=13)
    plt.ylabel('y',size=13)
    plt.title('Percorso migliore - '+mappa[i],size=15)
plt.show()

for i in range(2): print('Lunghezza del percorso migliore - '+mappa[i],fitness[i])
```
