## The Salesman Problem - Simulated Annealing

### Mappa della citt√†

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def CircleUp(x):
    return np.sqrt(1-x*x)
def CircleDown(x):
    return -np.sqrt(1-x*x)

f = plt.figure(figsize = (13,6))

x1,y1 = np.loadtxt('circle.start',usecols = (0,1), delimiter = ' ', unpack = 'true')
x2,y2 = np.loadtxt('square.start',usecols = (0,1), delimiter = ' ', unpack = 'true')

f.add_subplot(1, 2, 1)
x_c = np.linspace(-1,1,400)
plt.plot(x1,y1,'o')
plt.plot(x_c,CircleUp(x_c),color = 'green')
plt.plot(x_c,CircleDown(x_c),color='green')
plt.title('citt√†: disposizione random su circonferenza unitaria', size ='13')
plt.xlabel('x')
plt.ylabel('y')

f.add_subplot(1, 2, 2)
x_s = [0,0,1,1,0]
y_s = [0,1,1,0,0]
plt.plot(x2,y2,'o')
plt.plot(x_s,y_s,'-')
plt.title('citt√†: disposizione random nel quadrato [0,1)x[0,1)', size ='13')
plt.xlabel('x')
plt.ylabel('y')
#plt.xlim(0,1)
#plt.ylim(0,1)

plt.show()
```

## Fitness

```python
# fitness mediata sulla prima met√† della popolazione in funzione della generazione
# fitness del miglior individuo in funzione della generazione

f = plt.figure(figsize = (12,5))
temp,fitness = np.loadtxt("Circular/fitness.dat", usecols=(0,1), delimiter=' ', unpack='true')

f.add_subplot(1, 2, 1)
plt.plot(temp,fitness)
plt.xlabel('temperatura inversa')
plt.ylabel('fitness miglior individuo $L$')
plt.title('fitness miglior individuo vs generazione', size = '15')
plt.xlim(-5,130)
plt.grid()
plt.show()

print("Best path's fitness: ",fitness[999])
```

### Plot del percorso pi√π breve


```python
# plot del percorso pi√π breve 

ID = np.loadtxt('Circular/CirclePath.999.final',usecols = (0), delimiter = ' ', unpack = 'true')
ID = np.array(ID, dtype=int)

x = np.zeros(31)
y = np.zeros(31)

for i in range(30):
    x[i] = x1[ID[i]]
    y[i] = y1[ID[i]]
    
x[30] = x1[ID[0]]
y[30] = y1[ID[0]]

    
plt.figure(figsize = (6,6))
plt.plot(x,y,'o-')
plt.show()
    
```
